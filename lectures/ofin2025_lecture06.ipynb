### The Dividend-Discount Model

**A One-Year Investor**

Timeline:
```text
0         1
|---------|
-P₀       Div₁ + P₁
```

$$P_0 = \frac{Div_1 + P_1}{1 + r_E}$$

#### Dividend Yields, Capital Gains, and Total Returns

$$r_E = \frac{Div_1 + P_1}{P_0} - 1 = \underbrace{\frac{Div_1}{P_0}}_{\text{Dividend Yield}} + \underbrace{\frac{P_1 - P_0}{P_0}}_{\text{Capital Gain Rate}}$$

### The Dividend-Discount Model

* PentaCorp has a current stock price of $\$25$. It is expected to pay a dividend of $\$1.50$ in one year. Its expected price right after paying that dividend is $\$27$. Calculate PentaCorp's expected dividend yield, expected capital gain rate, and equity cost of capital.

**Expected Dividend Yield:**

$ \text{Dividend Yield} = \frac{Div_1}{P_0} \qquad \Rightarrow \qquad \text{Dividend Yield} = \frac{\$1.50}{\$25} = 0.06 \quad \text{or} \quad 6.00\%$

**Expected Capital Gain Rate:**

$ \text{Capital Gain Rate} = \frac{P_1 - P_0}{P_0} \qquad \Rightarrow \qquad \text{Capital Gain Rate} = \frac{\$27 - \$25}{\$25} = \frac{\$2}{\$25} = 0.08 \quad \text{or} \quad 8.00\%$

**Equity Cost of Capital ($r_E$):**
The equity cost of capital is the total expected return, which is the sum of the dividend yield and the capital gain rate.

$ r_E = \text{Dividend Yield} + \text{Capital Gain Rate} \qquad \Rightarrow \qquad r_E = 0.06 + 0.08 = 0.14 \quad \text{or} \quad 14.00\%$

* Expected Dividend Yield = 6.00%
* Expected Capital Gain Rate = 8.00%
* Equity Cost of Capital = 14.00%

### The Dividend-Discount Model

* Suppose AgeTech Corp. has a current stock price of $\$60$, is expected to pay a dividend of $\$2.40$ in one year, and its equity cost of capital is $12\%$. What price ($P_1$) must AgeTech stock sell for immediately after paying the dividend in one year to justify the current $\$60$ price?

    We start with the one-year investor valuation model:
    
    $ P_0 = \frac{Div_1 + P_1}{1 + r_E} $
    
    We are given $P_0 = \$60$, $Div_1 = \$2.40$, and $r_E = 12\% = 0.12$. We need to find $P_1$.

    Rearranging the formula to solve for $P_1$:
    
    $ P_1 = P_0 \times (1 + r_E) - Div_1 $
    
    $ P_1 = \$60 \times (1 + 0.12) - \$2.40 \qquad \Rightarrow \qquad P_1 = \$67.20 - \$2.40 = \$64.80 $

    Alternatively, using the total return decomposition:
    
    $ r_E = \underbrace{\frac{Div_1}{P_0}}_{\text{Dividend Yield}} + \underbrace{\frac{P_1 - P_0}{P_0}}_{\text{Capital Gain Rate}} \qquad \Rightarrow \qquad 12\% = \frac{\$2.40}{\$60} + \frac{P_1 - \$60}{\$60} \qquad \Rightarrow \qquad 0.08 = \frac{P_1 - \$60}{\$60} \qquad \Rightarrow \qquad P_1 = \$64.80 $

### The Dividend-Discount Model

* Suppose you expect Caspian Energy Corp. to pay dividends of $\$2.50$ per share and trade for $\$95.00$ per share at the end of the year. 
    * If investments with equivalent risk to Caspian Energy's stock have an expected return of $11\%$, what is the most you would pay today for Caspian Energy's stock? 
    * What dividend yield and capital gain rate would you expect at this price?

    The maximum price you would pay today ($P_0$) is the present value of the expected future cash flows (dividend and selling price):
    $$ P_0 = \frac{D_1 + P_1}{1 + r_E} \implies P_0 = \frac{2.50 + 95.00}{1 + 0.11} = \frac{97.50}{1.11} = \$87.84 $$
    
    At this price, the expected dividend yield is:
    $$ \text{Dividend Yield} = \frac{D_1}{P_0} = \frac{2.50}{87.84} = 0.0285 \text{ or } 2.85\% $$

    And the expected capital gain rate is:
    $$ \text{Capital Gain Rate} = \frac{P_1 - P_0}{P_0} = \frac{95.00 - 87.84}{87.84} = \frac{7.16}{87.84} = 0.0815 \text{ or } 8.15\% $$

### The Dividend-Discount Model

**A Multiyear Investor Timeline**

```text
0         1         2
|---------|---------|
-P₀       Div₁      Div₂ + P₂

```
$ P_0 = \frac{Div_1 + P_1}{1 + r_E} \qquad \text{and} \qquad P_1 = \frac{Div_2 + P_2}{1 + r_E} \qquad \Rightarrow \qquad P_0 = \frac{Div_1}{1 + r_E} + \frac{1}{1 + r_E} \left( \frac{Div_2 + P_2}{1 + r_E} \right) \qquad \Rightarrow \qquad P_0 = \frac{Div_1}{1 + r_E} + \frac{Div_2}{(1 + r_E)^2} + \frac{P_2}{(1 + r_E)^2} $

**The Dividend-Discount Model Equation**

$ P_0 = \frac{Div_1}{1 + r_E} + \frac{Div_2}{(1 + r_E)^2} + \dots + \frac{Div_N}{(1 + r_E)^N} + \frac{P_N}{(1 + r_E)^N} $

Generalizing for infinite horizon: $\qquad \qquad P_0 = \frac{Div_1}{1 + r_E} + \frac{Div_2}{(1 + r_E)^2} + \frac{Div_3}{(1 + r_E)^3} + \dots = \sum_{n=1}^{\infty} \frac{Div_n}{(1 + r_E)^n} $

For Perpetuity (Constant Dividend): $\qquad \quad P_0 = \frac{D_1}{r_E} $

**The price of the stock is equal to the present value of the expected future dividends it will pay.**

### The Dividend-Discount Model

* Suppose Beta Industries will pay a dividend of $\$3.50$ per share at the end of this year and $\$3.75$ per share next year. You expect Beta Industries' stock price to be $\$62.50$ in two years. If Beta Industries' equity cost of capital is $9.5\%$, what price ($P_0$) would you be willing to pay for a share of its stock today, if you planned to hold the stock for two years?

    $ P_0 = \frac{D_1}{(1 + r_E)^1} + \frac{D_2 + P_2}{(1 + r_E)^2} \qquad \Rightarrow \qquad P_0 = \frac{3.50}{(1 + 0.095)^1} + \frac{3.75 + 62.50}{(1 + 0.095)^2} \qquad \Rightarrow \qquad P_0 = \$58.45 $

* Suppose instead you plan to hold the stock for only one year. What price ($P_1$) would you expect to be able to sell a share of Beta Industries stock for in one year?

    $ P_1 = \frac{D_2 + P_2}{(1 + r_E)^1} \qquad \Rightarrow \qquad P_1 = \frac{3.75 + 62.50}{(1 + 0.095)} \qquad \Rightarrow \qquad P_1 = \$60.50 $

* Given your answer above, what price ($P_0$) would you be willing to pay for a share of Beta Industries stock today, if you planned to hold the stock for one year?

    $ P_0 = \frac{D_1 + P_1}{(1 + r_E)^1} \qquad \Rightarrow \qquad P_0 = \frac{3.50 + 60.50}{(1 + 0.095)}  \qquad \Rightarrow \qquad P_0 = \$58.45 $

### The Dividend-Discount Model: Constant Dividend Growth

Timeline:
```text
0         1         2             3             ...
|---------|---------|-------------|-------------
-P₀       Div₁      Div₁(1+g)     Div₁(1+g)²   ...
```

$$ P_0 = \frac{D_1}{r_E - g} $$

Rearranging for $r_E$:

$$ r_E = \frac{D_1}{P_0} + g $$

A constant growth perpetuity:

$$ P_0 = \frac{D_1}{r_E - g} $$

With constant expected dividend growth, the expected growth rate of the share price matches the growth rate of dividends.

    $P_0 = \frac{D_1}{r_E - g} = \frac{\$0.84}{0.10 - 0.05} = \frac{\$0.84}{0.05} = \$16.80$

**What is the expected growth rate of Arturu's share price?**

Assuming constant dividend growth, the share price is expected to grow at the same rate as the dividends. Therefore, the expected growth rate of the share price is also $6.4\%$.

### Dividends Versus Investment and Growth

$$ P_0 = \frac{D_1}{r_E - g} $$

According to the constant dividend growth model, for a given cost of capital, the firm's share price increases with the current dividend level, and the expected growth rate.


**A Simple Model of Growth**

$$ Div_t = \underbrace{\frac{\text{Earnings}_t}{\text{Shares Outstanding}_t}}_{\text{EPS}_t} \times \text{Dividend Payout Rate}_t $$

**Dividend Payout Rate:** The fraction of its earnings that a firm pays as dividends each year.

$ \text{Change in Earnings} = \text{New Investment} * \text{Return on New Investment} \qquad \text{and} \qquad \text{New Investment} = \text{Earnings} * \text{Retention Rate} $

$ \text{Change in Earnings} = \text{Earnings} * \text{Retention Rate} * \text{Return on New Investment} $

$ \frac{\text{Change in Earnings}}{\text{Earnings}} = \text{Retention Rate} \times \text{Return on New Investment} $

$ g = \text{Retention Rate} \times \text{Return on New Investment} $

**Sustainable Growth Rate:** The rate at which a firm can grow using only retained earnings.

### Dividends Versus Investment and Growth

PSK Industries expects to have earnings per share (EPS) of $\$8$ in the coming year. The firm currently plans to pay out all earnings as dividends, resulting in no growth. With these expectations, PSK's current share price is $\$80$.

Suppose PSK could reduce its dividend payout rate to $60\%$ permanently and use the retained earnings to invest in new projects. The return on investment (RONI) for these projects is expected to be $15\%$. Assuming PSK's equity cost of capital remains unchanged, what effect would this new dividend policy have on its stock price?

**Original Policy (No Growth):**

The dividend next year ($D_1$) equals EPS, so $D_1 = \$8$. The growth rate ($g$) is 0.

$ P_0 = \frac{D_1}{r_E - g} \qquad \Rightarrow \qquad \$80 = \frac{\$8}{r_E - 0} \qquad \Rightarrow \qquad r_E = \frac{\$8}{\$80} = 0.10 = 10\% $

**New Policy (Growth):**

$\text{New Payout Rate} = 60\% \qquad \Rightarrow \qquad \text{Retention Rate} = 1 - 0.60 = 0.40 \quad \text{or} \quad 40\%$

*Return on New Investment (RONI) = 15%.*

$ g' = \text{Retention Rate} \times \text{RONI} = 0.40 \times 15\% = 0.06 = 6\%$

$ D'_1 = EPS_1 \times \text{New Payout Rate} = \$8 \times 60\% = \$4.80 $

$ P'_0 = \frac{D'_1}{r_E - g'} = \frac{\$4.80}{10\% - 6\%} = \frac{\$4.80}{0.04} = \$120 $

---

### Dividends Versus Investment and Growth

Now suppose that the return on new investments for PSK Industries is 9%, rather than 15%. 
    
* Given its expected EPS of $\$8$ and equity cost of capital of $10\%$, what will happen to PSK's current share price if it adopts the $60\%$ payout policy?

**New Policy (Lower RONI):**

*$\text{Retention Rate} = 40\%$.*

*$\text{Return on New Investment (RONI)} = 9\%$.*

$ g'' = \text{Retention Rate} \times \text{RONI} = 0.40 \times 9\% = 0.036 = 3.6\% $

$D''_1 = \$8 \times 60\% = \$4.80$.

$ P''_0 = \frac{D''_1}{r_E - g''} = \frac{\$4.80}{10\% - 3.6\%} = \frac{\$4.80}{0.064} = \$75 $

Reducing the dividend payout to fund investments will increase the firm's stock price only if the return on the new investments (RONI) is greater than the equity cost of capital ($r_E$). 

* In the first case, $15\% > 10\%$, so the price increased. 

* In the second case, $9\% < 10\%$, so the price decreased. 

**Investing in projects with a return lower than the cost of capital destroys value.**

### Dividends Versus Investment and Growth

BSM Industries just announced it will cut its dividend from $\$3.50$ to $\$2.00$ per share and use the extra funds to expand its operations. 

* Prior to the announcement, BSM's dividends were expected to grow at a $3.0\%$ rate indefinitely, and its share price was $\$50.00$. 

* With the planned expansion, BSM's dividends are expected to grow at a $5.5\%$ rate indefinitely. 

* What share price would you expect immediately after the announcement? (Assume BSM's risk, and therefore its equity cost of capital, is unchanged by the expansion.) Based on this share price, does the expansion appear to be a positive NPV investment?


$P_0 = \$50.00$, $D_1 = \$3.50$, $g = 3.0\%$

$ P_0 = \frac{D_1}{r_E - g} \qquad \Rightarrow \qquad \$50.00 = \frac{\$3.50}{r_E - 0.03} \qquad \Rightarrow \qquad r_E = 0.10 \quad \text{or} \quad 10.0\% $
    

$D'_1 = \$2.00$, $g' = 5.5\%$
   
The equity cost of capital ($r_E$) remains 10.0% as risk is assumed unchanged.

$ P'_0 = \frac{D'_1}{r_E - g'} \qquad \Rightarrow \qquad P'_0 = \frac{\$2.00}{0.10 - 0.055} \qquad \Rightarrow \qquad P'_0 \approx \$44.44 $

$P'_0 \approx \$44.44 < P_0 = \$50.00$.

**The expansion appears to be a negative NPV investment and the return generated by the funds reinvested into the expansion is implicitly less than the required return ($r_E = 10.0\%$).**

$$ P_0 = \frac{Div_1}{1 + r_E} + \frac{Div_2}{(1 + r_E)^2} + \dots + \frac{Div_N}{(1 + r_E)^N} + \frac{1}{(1 + r_E)^N} \left( \frac{Div_{N+1}}{r_E - g} \right) $$

### Stock Valuation: Changing Growth Rates

* Arturu Corp. has developed a novel AI-driven logistics optimization software. 
    * The market reaction has been strong, and Arturu is reinvesting all earnings to scale up. Earnings were $\$3$ per share this past year (Year 0) and are projected to grow at $25\%$ per year through the end of Year 3. 
    * After Year 3, competition is expected to increase. Analysts predict that Arturu will then reduce investment, begin paying out $50\%$ of its earnings as dividends, and growth will stabilize at a long-run rate of $5\%$ per year. 
    * If Arturu's equity cost of capital ($r_E$) is $10\%$, what is the estimated value of a share today ($P_0$)?

**Projected Earnings and Dividends:**

| Year                  | 0       | 1       | 2       | 3       | 4       | 5       |
| :-------------------- | ------: | ------: | ------: | ------: | ------: | ------: |
| **Earnings** |         |         |         |         |         |         |
| 1 EPS Growth Rate (g) |         | 25%     | 25%     | 25%     | 5%      | 5%      |
| 2 EPS (\$)             | 3.00   | 3.75   | 4.69   | 5.86   | 6.15   | 6.46   |
| **Dividends** |         |         |         |         |         |         |
| 3 Dividend Payout Rate|         | 0%      | 0%      | 0%      | 50%     | 50%     |
| 4 Dividend (Div) (\$)  |         | 0.00   | 0.00   | 0.00   | 3.08   | 3.23   |

1.  **Calculate the Terminal Value ($P_3$):** The terminal value represents the value of all dividends from Year 4 onwards, discounted back to the end of the high-growth period (Year 3). We use the constant growth dividend discount model (Gordon Growth Model) using the Year 4 dividend.

    $ P_3 = \frac{Div_4}{r_E - g_{\text{long-run}}} \qquad \Rightarrow \qquad P_3 = \frac{3.076}{0.10 - 0.05} \approx \$61.52 $

2.  **Calculate the Present Value ($P_0$):** The value today is the present value of the dividends during the initial high-growth phase (which are zero in this case) plus the present value of the terminal value $P_3$.
    
    $ P_0 = \frac{Div_1}{(1 + r_E)^1} + \frac{Div_2}{(1 + r_E)^2} + \frac{Div_3 + P_3}{(1 + r_E)^3} $
    
    Since $Div_1, Div_2, Div_3$ are all $0$:
    $ P_0 = \frac{0}{(1.10)^1} + \frac{0}{(1.10)^2} + \frac{0 + 61.52}{(1.10)^3} \qquad \Rightarrow \qquad P_0 \approx \$46.22 $

### Stock Valuation: Changing Growth Rates

* Arturu Corp (ART) paid an annual dividend of $\$1.50$ in 2024. 
    * You expect ART to increase its dividends by $10\%$ per year for the next four years (through 2028), and thereafter by $4\%$ per year indefinitely. 
    * If the appropriate equity cost of capital for Arturu Corp is $9\%$ per year, use the dividend-discount model to estimate its value per share at the end of 2024.

    $D_{2025} = D_{2024} \times (1 + g_1) = 1.50 \times (1.10)^1 = 1.65$

    $D_{2026} = D_{2024} \times (1.10)^2 = 1.50 \times 1.21 = 1.815$

    $D_{2027} = D_{2024} \times (1.10)^3 = 1.50 \times 1.331 = 1.9965$

    $D_{2028} = D_{2024} \times (1.10)^4 = 1.50 \times 1.4641 = 2.19615$
    
    $D_{2029} = D_{2028} \times (1 + g_2) = 2.19615 \times (1.04) = 2.283996$

    *Calculate the Terminal Value at the end of the high-growth phase (end of 2028, $P_{2028}$):*
    
    $ P_{2028} = \frac{D_{2029}}{r_E - g_2} \qquad \Rightarrow \qquad P_{2028} = \frac{2.283996}{0.09 - 0.04} = \$45.67992 $

    *Calculate the Value per share at the end of 2024 ($P_{2024}$):*
    
    $ P_{2024} = \frac{D_{2025}}{(1+r_E)^1} + \frac{D_{2026}}{(1+r_E)^2} + \frac{D_{2027}}{(1+r_E)^3} + \frac{D_{2028} + P_{2028}}{(1+r_E)^4} $

    $ P_{2024} = \frac{1.65}{(1.09)^1} + \frac{1.815}{(1.09)^2} + \frac{1.9965}{(1.09)^3} + \frac{2.19615 + 45.67992}{(1.09)^4} $
    
    $ P_{2024} \approx \$38.50 $

````python
# If you don’t already have ipywidgets installed, uncomment:
# !pip install ipywidgets

import numpy as np
import ipywidgets as widgets
from IPython.display import display, clear_output

def recalc_ddm(_):
    with output:
        clear_output()
        # Read inputs using "initial" terminology
        base_year       = year0_widget.value
        D0              = D0_widget.value
        g_initial       = g_initial_widget.value # Renamed from g_high
        g_stable        = g2_widget.value
        years_initial   = int(n_initial_widget.value) # Renamed from years_high
        r_initial       = r_initial_widget.value # Renamed from r_high
        r_stable        = r_stable_widget.value

        # --- Calculations ---
        # Input validation
        if r_stable <= g_stable:
             # Display rates with 2 decimal places as percentages in error messages
             print(f"\nError: Stable discount rate ({r_stable:.2%}) must be greater than stable growth rate ({g_stable:.2%}).")
             return
        if r_initial <= -1: # Avoid issues with (1+r_initial) being zero or negative
             print(f"\nError: Initial discount rate (r_initial={r_initial:.2%}) must be greater than -100%.")
             return

        # 1) Initial-growth phase dividends (D1 to DN)
        dividends = [D0 * (1 + g_initial)**n for n in range(1, years_initial + 1)]

        # 2) First stable-growth dividend (D_{N+1})
        # Handle case where years_initial might be 0 (though widget prevents this)
        last_initial_growth_div = dividends[-1] if years_initial > 0 else D0 * (1 + g_initial)
        D_stable = last_initial_growth_div * (1 + g_stable)

        # 3) Terminal value at end of initial growth (P_N, using stable rate in formula)
        P_terminal = D_stable / (r_stable - g_stable) # Renamed P_high to P_terminal

        # 4) PV of initial-growth dividends (discounted at r_initial)
        pv_divs = sum(dividends[i] / (1 + r_initial)**(i+1) for i in range(years_initial))

        # 5) PV of terminal value ONLY (discounted at r_initial)
        pv_terminal_only = P_terminal / (1 + r_initial)**years_initial

        # 6) Total value
        P0 = pv_divs + pv_terminal_only

        # --- Display ---
        # Format output values to 4 decimal places for precision, except P0 to 2
        print(f"Initial-growth phase dividends:")
        for i, div in enumerate(dividends, 1):
            yr = base_year + i
            print(f"  D_{yr} = {div:.4f}") # Keep dividend precision higher
        print(f"\nFirst Stable-growth dividend (D_{base_year + years_initial + 1}) = {D_stable:.4f}")
        print(f"Terminal value at end of initial growth (P_{base_year + years_initial}) = {P_terminal:.4f}")
        print(f"\nEstimated share value at end of {base_year} (P₀): ${P0:.2f}") # Display final price with 2 decimals

# Widget styling
style = {'description_width': '200px'}
layout = widgets.Layout(width='380px') # Slightly wider for longer descriptions

# Widgets (using "initial" terminology and .2f format)
year0_widget = widgets.BoundedIntText(
    value=2024, min=1900, max=2100, step=1,
    description='Base year for D₀:', style=style, layout=layout)
D0_widget = widgets.FloatText(
    value=1.50, step=0.01, # Step controls increment, implicitly suggests 2 decimals
    description='D₀ (Dividend in Base Year):', style=style, layout=layout)
g_initial_widget = widgets.FloatSlider(
    value=0.10, min=-0.10, max=0.30, step=0.005, readout_format='.2f', # Set readout format
    description='Initial growth rate (g₁):', style=style, layout=layout)
n_initial_widget = widgets.BoundedIntText(
    value=4, min=1, max=20, step=1,
    description='Initial growth years (N):', style=style, layout=layout)
g2_widget = widgets.FloatSlider(
    value=0.04, min=-0.10, max=0.15, step=0.001, readout_format='.2f', # Set readout format
    description='Stable growth rate (g₂):', style=style, layout=layout)
r_initial_widget = widgets.FloatSlider(
    value=0.09, min=0.00, max=0.25, step=0.001, readout_format='.2f', # Set readout format
    description='Discount rate (initial, r₁):', style=style, layout=layout)
r_stable_widget = widgets.FloatSlider(
    value=0.09, min=0.00, max=0.25, step=0.001, readout_format='.2f', # Set readout format
    description='Discount rate (stable, r₂):', style=style, layout=layout)

recalc_button = widgets.Button(
    description='Recalculate Value',
    button_style='info',
    layout=widgets.Layout(width='220px'))
recalc_button.on_click(recalc_ddm)

output = widgets.Output()

# Layout (using new widget names)
ui = widgets.VBox([
    year0_widget,
    D0_widget,
    g_initial_widget,
    n_initial_widget,
    g2_widget,
    r_initial_widget,
    r_stable_widget,
    recalc_button,
    output
])
display(ui)

# Initial run
recalc_ddm(None)
````

### Limitations of the Dividend-Discount Model
*   There is a large amount of uncertainty associated with any forecast of a firm's future dividends.
*   Price estimations are sensitive to dividend growth assumptions and even small changes in the assumed dividend growth rate can lead to significant changes in these estimations.
*   Forecasting dividends requires forecasting the firm's earnings, dividend payout ratio, and future share count, which are at management's discretion.

### Total Payout and Free Cash Flow Valuation Models

**Share Repurchase:** A firm using its excess cash to buy back its own stock.
*   Two consequences for the dividend-discount model.
    *   The more cash a firm uses to repurchase shares, the less it has available to pay dividends.
    *   By repurchasing shares, the firm decreases its share count, which increases its earnings and dividends on a per-share basis.

**Total Payout Model** values the equity of the firm rather than a single share.

$$ \text{Total Firm Value} = PV(\text{Future Total Dividends and Repurchases}) $$

$$ P_0 = \frac{PV(\text{Future Total Dividends and Repurchases})}{\text{Shares Outstanding}_0} $$

*   Total payout model values all of the firm's equity, rather than a single share.
*   **Discount total dividends and share repurchases and use the growth rate of total earnings (rather than earnings per share) when forecasting the growth of the firm's total payouts.**

### Total Payout and Free Cash Flow Valuation Models

* Arturu Corporation has $150 \, \text{million}$ shares outstanding and expects earnings at the end of this year of $\$600 \, \text{million}$. Arturu plans to pay out $60\%$ of its earnings in total, paying $40\%$ as a dividend and using $20\%$ to repurchase shares. If Arturu's earnings are expected to grow by $6\%$ per year and these payout rates remain constant, determine Arturu's share price assuming an equity cost of capital ($r_E$) of $9\%$.

    Total Payouts (Year 1) = $60\% \times \$600 \text{ million} = \$360 \text{ million}$.
    
    Growth rate of payouts = Growth rate of earnings = $g = 6\%$.
    
    Equity cost of capital = $r_E = 9\%$.

    Total Value of Equity using the constant growth total payout model:
    $$ \text{Total Value} = \frac{\text{Total Payouts}_1}{r_E - g} = \frac{\$360 \text{ million}}{0.09 - 0.06} = \frac{\$360 \text{ million}}{0.03} = \$12,000 \text{ million} = \$12 \text{ billion} $$

    Share Price ($P_0$):
    $$ P_0 = \frac{\text{Total Value}}{\text{Shares Outstanding}_0} = \frac{\$12,000 \text{ million}}{150 \text{ million shares}} = \$80.00 \text{ per share} $$

### Total Payout and Free Cash Flow Valuation Models

* Arturu Corporation has $150 \, \text{million}$ shares outstanding and expects earnings at the end of this year of $\$600 \, \text{million}$. Arturu plans to pay out $60\%$ of its earnings in total, paying $40\%$ as a dividend and using $20\%$ to repurchase shares.     Total Payouts (Year 1) = $60\% \times \$600 \text{ million} = \$360 \text{ million}$.
    
    Growth rate of payouts = Growth rate of earnings = $g = 6\%$.
    
    Equity cost of capital = $r_E = 9\%$.
    
    Dividend per share (Year 1): $\quad Div_1 = \frac{\text{Dividend Payouts}_1}{\text{Shares Outstanding}_0} \quad \Rightarrow \quad Div_1 = \frac{40\% \times \$600 \text{ million}}{150 \text{ million shares}} = \frac{\$240 \text{ million}}{150 \text{ million}} = \$1.60 \text{ per share} $

    $\text{Since} \quad r_E = \text{Dividend Yield} + \text{Capital Gain Rate} \quad \Rightarrow \quad \text{Capital Gain Rate} = r_E - \text{Dividend Yield} = 9.00\% - 2.00\% = 7.00\% $

    * Because the company repurchases shares (using $20\%$ of earnings), the number of shares decreases over time. 
        * This causes the earnings per share (EPS) and dividends per share (DPS) to grow faster than total earnings, and results in a capital gain rate $(7.00\%)$ that is higher than the overall growth rate of the firm's total payouts $(6\%)$. 
        * The total return $r_E=9\%$ is composed of the dividend yield $(2\%)$ and the capital gain $(7\%)$.

### The Discounted Free Cash Flow Model

*   **Market Capitalization:** The market value of the firm's equity, or the value that remains after the firm has paid its debts.

*   **Enterprise Value:** The value of the underlying business assets, unencumbered by debt and separate from any cash and marketable securities.
    *   The net cost of acquiring the firm's equity, taking its cash, paying off all debt, and thus owning the unlevered business.

$$ \text{Enterprise Value} = \text{Market Value of Equity} + \text{Debt} - \text{Cash} $$

| Present Value of ...                 | At the ...                       | Determines the ... |
| :----------------------------------- | :------------------------------- | :----------------- |
| Dividend Payments                    | Equity cost of capital           | Stock Price        |
| Total Payouts (All dividends and repurchases) | Equity cost of capital           | Equity Value       |
| Free Cash Flow (Cash available to pay all security holders) | Weighted average cost of capital | Enterprise Value   |

*   To estimate a firm's enterprise value, we compute the present value of the free cash flow (FCF) that the firm has available to pay all investors, both debt and equity holders.

---

### The Discounted Free Cash Flow Model

$ \text{Free Cash Flow} = \overbrace{(\text{Revenues} - \text{Costs} - \text{Depreciation})}^{\text{EBIT}} \times (1 - \tau_c) + \text{Depreciation} - \text{CapEx} - \Delta NWC $

$\qquad \text{where} (EBIT \times (1 - \tau_c)) \quad \text{is Unlevered Net Income}$

$ \text{Free Cash Flow} = \underbrace{\text{EBIT} \times (1 - \tau_c)}_{\text{Unlevered Net Income}} + \text{Depreciation} - \text{CapEx} - \Delta NWC $

$ \text{Free Cash Flow} = \underbrace{\text{EBIT} \times (1 - \tau_c)}_{\text{Unlevered Net Income}} - \text{Net Investment} - \Delta NWC $

$\qquad \text{where} \quad \text{Net Investment} = \text{Capital Expenditures} - \text{Depreciation} $

**Enterprise Value ($V_0$) = PV(Future Free Cash Flow of Firm)**

$ V_0 = PV (\text{Future Free Cash Flow of Firm})$  

$EV = \text{Market Value of Equity} + \text{Debt} - \text{Cash} \qquad \Rightarrow \qquad \text{Market Value of Equity} = EV - \text{Debt} + \text{Cash} $

$ P_0 = \frac{\text{Market Value of Equity}}{\text{Shares Outstanding}_0} = \frac{EV - \text{Debt}_0 + \text{Cash}_0}{\text{Shares Outstanding}_0} \qquad \text{or} \qquad P_0 = \frac{V_0 + \text{Cash}_0 - \text{Debt}_0}{\text{Shares Outstanding}_0} $

*A firm's free cash flow is equal to the sum of the free cash flows from the firm's current and future investments. Therefore, we can interpret the firm's enterprise value as the total NPV that the firm will earn from continuing its existing projects and initiating new ones.*

### Discount Rate in DCF Model
**Weighted average cost of capital (WACC)**
    
*   The average cost of capital the firm must pay to all of its investors, both debt and equity holders.
    
*   If the firm has no debt, then $r_{wacc} = r_E$.
    
*   When a firm has debt, $r_{wacc}$ is a weighted average of the firm's debt and equity cost of capital.
    
*   Since debt is generally less risky than equity, $r_{wacc}$ is generally less than $r_E$.
    
*   The WACC reflects the average risk of all of the firm's investments

**Valuation Formula:**

$ V_0 = \frac{FCF_1}{1 + r_{wacc}} + \frac{FCF_2}{(1 + r_{wacc})^2} + \dots + \frac{FCF_N}{(1 + r_{wacc})^N} + \frac{V_N}{(1 + r_{wacc})^N} \qquad \text{where} \quad V_N \quad \text{is the Terminal Enterprise Value at Year N.}$


Terminal Value assuming constant FCF growth $g_{FCF}$ beyond year N:

$ V_N = \frac{FCF_{N+1}}{r_{wacc} - g_{FCF}} = \left( \frac{1 + g_{FCF}}{r_{wacc} - g_{FCF}} \right) \times FCF_N \qquad \text{where} \quad g_{FCF} \quad \text{is the long-term FCF growth rate}$

### DCF Valuation
AGE Corp had sales of $\$600 \, \text{million}$ in 2024. Suppose you expect its sales to grow at an $8\%$ rate in 2025, but that this growth rate will slow by $1\%$ per year to a long-run growth rate for its industry of $3.5\%$ by 2030. Based on AGE Corp's past profitability and investment needs, you expect EBIT to be $10\%$ of sales, net investment to be $7\%$ of any increase in sales, and increases in net working capital requirements to be $12\%$ of any increase in sales.

* If AGE Corp has $\$80 \, \text{million}$ in cash, $\$10\, \text{million}$ in debt, $30\, \text{million}$ shares outstanding, a tax rate ($\tau$) of $35\%$, and a weighted average cost of capital ($r_{wacc}$) of $10\%$, what is your estimate of the value of AGE Corp's stock in early 2025?

**1. Free Cash Flow Forecast ($ millions)**

| Line | Item                  | Formula                  | 2024  | 2025    | 2026    | 2027    | 2028    | 2029    | 2030    |
| :--- | :-------------------- | :----------------------- | :---- | :------ | :------ | :------ | :------ | :------ | :------ |
| 9    | **Free Cash Flow (FCF)**| (6) - (7) - (8)          |       | **32.90** | **36.45** | **39.86** | **43.18** | **46.30** | **48.66** |


**2. Terminal Value Calculation**

The terminal value ($V_{2029}$) captures the value of FCF beyond the explicit forecast horizon (2029), assuming a constant growth rate ($g$) of 3.5% thereafter. The calculation uses the FCF in the first year of the constant growth period ($FCF_{2030}$).

$$ V_{2029} = \frac{FCF_{2030}}{r_{wacc} - g} = \frac{48.66}{0.10 - 0.035} = \frac{48.66}{0.065} \approx \$748.62 \text{ million} $$

**3. Enterprise Value Calculation**

$ V_0 = \sum_{t=1}^{5} \frac{FCF_{2024+t}}{(1+r_{wacc})^t} + \frac{V_{2029}}{(1+r_{wacc})^5} \qquad \Rightarrow \qquad V_0 = \frac{32.90}{1.10^1} + \frac{36.45}{1.10^2} + \frac{39.86}{1.10^3} + \frac{43.18}{1.10^4} + \frac{46.30 + 748.62}{1.10^5} \qquad \Rightarrow \qquad V_0 \approx \$613.05 \text{ million} $

**4. Stock Price Estimation**

$ P_0 = \frac{V_0 + \text{Cash}_0 - \text{Debt}_0}{\text{Shares Outstanding}_0} = \frac{613.05 + 80 - 10}{30} = \frac{683.05}{30} \approx \$22.77 $

````python
# If you don’t already have ipywidgets installed, uncomment and run:
# !pip install ipywidgets

import numpy as np
import pandas as pd
import ipywidgets as widgets
from IPython.display import display, clear_output, HTML

# ─── Core Calculation Function ───────────────────────────────────────────────
def calculate_fcf_valuation(
    sales_2024, start_growth, growth_decline, stable_growth,
    stable_growth_year, ebit_margin, netinv_pct_delta_sales,
    nwc_pct_delta_sales, tax_rate, rwacc,
    cash, debt, shares_out):
    """
    Performs the multi-stage FCFF valuation based on inputs.
    Returns: DataFrame, Terminal Value, Enterprise Value, Stock Price, Error Msg,
             FCF_TerminalYear, PV_FCFs, PV_Terminal, Equity_Value
    """

    # --- Setup Years and Growth Rates ---
    forecast_start_year = 2025 # Year 1 of forecast
    last_explicit_year = stable_growth_year - 1
    forecast_years = list(range(forecast_start_year, stable_growth_year)) # Explicit forecast period
    first_stable_year = stable_growth_year

    growth_rates = {}
    current_growth = start_growth
    for year in forecast_years:
        growth_rates[year] = current_growth
        current_growth -= growth_decline
        # Ensure growth doesn't fall below stable rate during explicit period if decline is steep
        current_growth = max(current_growth, stable_growth)

    growth_rates[first_stable_year] = stable_growth

    # --- Project Financials ---
    forecast_data = {}
    last_sales = sales_2024

    # Calculate for explicit years + first stable year (needed for TV)
    all_calc_years = list(range(forecast_start_year, first_stable_year + 1))

    for year in all_calc_years:
        data = {}
        data['Growth Rate'] = growth_rates[year]
        data['Sales'] = last_sales * (1 + data['Growth Rate'])
        data['Delta Sales'] = data['Sales'] - last_sales
        data['EBIT'] = data['Sales'] * ebit_margin
        data['Tax'] = data['EBIT'] * tax_rate
        data['Unlevered NI'] = data['EBIT'] * (1 - tax_rate)
        delta_sales_for_inv = max(0, data['Delta Sales'])
        data['Net Investment'] = delta_sales_for_inv * netinv_pct_delta_sales
        data['Delta NWC'] = delta_sales_for_inv * nwc_pct_delta_sales
        data['FCF'] = data['Unlevered NI'] - data['Net Investment'] - data['Delta NWC']

        forecast_data[year] = data
        last_sales = data['Sales']

    # --- Create DataFrame for Display ---
    df_forecast = pd.DataFrame(forecast_data).T # Transpose for years as rows
    # Keep only explicit years for main display table, but keep 2030 data for TV calc
    df_display = df_forecast.loc[forecast_years].copy()

    # --- Terminal Value Calculation ---
    # Check if first_stable_year data exists before accessing
    if first_stable_year not in forecast_data:
         error_message = f"Error: Data for stable growth year {first_stable_year} could not be calculated."
         return df_display, np.nan, np.nan, np.nan, error_message, np.nan, np.nan, np.nan, np.nan

    fcf_terminal_year = forecast_data[first_stable_year]['FCF']
    g = stable_growth
    if rwacc <= g:
        # Handle invalid input case
        terminal_value_N = np.nan
        enterprise_value_0 = np.nan
        stock_price_0 = np.nan
        error_message = f"Error: WACC ({rwacc:.2%}) must be greater than Stable Growth Rate ({g:.2%})"
        # Add nan placeholders for the new return values
        return df_display, terminal_value_N, enterprise_value_0, stock_price_0, error_message, fcf_terminal_year, np.nan, np.nan, np.nan # Return calculated FCF even in error if possible
    else:
        error_message = None
        # V_N = FCF_{N+1} / (rwacc - g)
        terminal_value_N = fcf_terminal_year / (rwacc - g) # This is V_2029

        # --- Enterprise Value Calculation ---
        pv_fcf = 0
        # Discount explicit FCFs
        for i, year in enumerate(forecast_years):
             # Check if FCF data exists before accessing
             if year not in forecast_data or 'FCF' not in forecast_data[year] or pd.isnull(forecast_data[year]['FCF']):
                 error_message = f"Error: Missing or invalid FCF data for year {year}."
                 return df_display, terminal_value_N, np.nan, np.nan, error_message, fcf_terminal_year, np.nan, np.nan, np.nan
             pv_fcf += forecast_data[year]['FCF'] / ((1 + rwacc)**(i + 1))


        # Discount Terminal Value (V_N at end of year N=last_explicit_year)
        n_explicit = len(forecast_years) # Number of explicit FCF years
        pv_terminal = terminal_value_N / ((1 + rwacc)**n_explicit)

        enterprise_value_0 = pv_fcf + pv_terminal

        # --- Stock Price ---
        equity_value_0 = enterprise_value_0 + cash - debt
        stock_price_0 = equity_value_0 / shares_out if shares_out > 0 else np.nan

        # ADD fcf_terminal_year, pv_fcf, pv_terminal, equity_value_0 to return
        return df_display, terminal_value_N, enterprise_value_0, stock_price_0, error_message, fcf_terminal_year, pv_fcf, pv_terminal, equity_value_0


# ─── Alternative Scenario Calculation Functions ─────────────────────────────
# (Keep these functions as they were, they don't need the extra return values)
def calculate_constant_growth_scenario(
    sales_2024, stable_growth, ebit_margin, netinv_pct_delta_sales,
    nwc_pct_delta_sales, tax_rate, rwacc,
    cash, debt, shares_out):
    """Calculates valuation assuming constant growth from year 1."""
    g = stable_growth
    if rwacc <= g:
        return np.nan, np.nan, f"Error: WACC ({rwacc:.2%}) <= Stable Growth ({g:.2%})"

    # Calculate FCF for Year 1 (2025)
    sales_2025 = sales_2024 * (1 + g)
    delta_sales_2025 = sales_2025 - sales_2024
    ebit_2025 = sales_2025 * ebit_margin
    unlevered_ni_2025 = ebit_2025 * (1 - tax_rate)
    delta_sales_for_inv = max(0, delta_sales_2025)
    net_inv_2025 = delta_sales_for_inv * netinv_pct_delta_sales
    delta_nwc_2025 = delta_sales_for_inv * nwc_pct_delta_sales
    fcf_2025 = unlevered_ni_2025 - net_inv_2025 - delta_nwc_2025

    # Enterprise Value using Gordon Growth
    enterprise_value_0 = fcf_2025 / (rwacc - g)

    # Stock Price
    equity_value_0 = enterprise_value_0 + cash - debt
    stock_price_0 = equity_value_0 / shares_out if shares_out > 0 else np.nan

    return enterprise_value_0, stock_price_0, None # V0, P0, error message

def calculate_constant_growth_low_margin_scenario(
    sales_2024, stable_growth, low_ebit_margin, netinv_pct_delta_sales,
    nwc_pct_delta_sales, tax_rate, rwacc,
    cash, debt, shares_out):
    """Calculates valuation assuming constant growth and lower margin from year 1."""
    # Re-use the constant growth function with the lower margin
    return calculate_constant_growth_scenario(
        sales_2024, stable_growth, low_ebit_margin, netinv_pct_delta_sales,
        nwc_pct_delta_sales, tax_rate, rwacc,
        cash, debt, shares_out
    )


# ─── Widget Definitions ─────────────────────────────────────────────────────
style = {'description_width': 'initial'} # Auto-adjust description width
layout = widgets.Layout(width='350px')
layout_wide = widgets.Layout(width='500px') # For longer descriptions

# --- Input Widgets ---
w_sales_2024 = widgets.FloatText(value=600.0, description='Sales 2024 ($M):', style=style, layout=layout)
w_start_growth = widgets.FloatText(value=0.08, step=0.005, description='Growth Rate 2025:', style=style, layout=layout)
w_growth_decline = widgets.FloatText(value=0.01, step=0.001, description='Annual Growth Decline:', style=style, layout=layout)
w_stable_growth = widgets.FloatText(value=0.035, step=0.001, description='Stable Growth Rate (g):', style=style, layout=layout)
w_stable_growth_year = widgets.IntText(value=2030, step=1, description='Year Stable Growth Begins:', style=style, layout=layout)
w_ebit_margin = widgets.FloatText(value=0.10, step=0.005, description='EBIT Margin (% of Sales):', style=style, layout=layout)
w_netinv_pct = widgets.FloatText(value=0.07, step=0.005, description='Net Inv (% of ΔSales):', style=style, layout=layout)
w_nwc_pct = widgets.FloatText(value=0.12, step=0.005, description='ΔNWC (% of ΔSales):', style=style, layout=layout)
w_tax_rate = widgets.FloatText(value=0.35, step=0.01, description='Tax Rate (τ):', style=style, layout=layout)
w_rwacc = widgets.FloatText(value=0.10, step=0.001, description='WACC (rwacc):', style=style, layout=layout)
w_cash = widgets.FloatText(value=80.0, description='Cash ($M):', style=style, layout=layout)
w_debt = widgets.FloatText(value=10.0, description='Debt ($M):', style=style, layout=layout)
w_shares_out = widgets.FloatText(value=30.0, description='Shares Out (Millions):', style=style, layout=layout)

# --- Button and Output ---
recalc_button = widgets.Button(description='Calculate Valuation', button_style='info', layout=widgets.Layout(width='200px'))
output_area = widgets.Output()

# ─── Display Logic ──────────────────────────────────────────────────────────
def display_valuation(_): # Triggered by button
    with output_area:
        clear_output(wait=True)

        # --- Read Values ---
        # Assigning to local variables for readability
        sales_2024 = w_sales_2024.value
        start_growth = w_start_growth.value
        growth_decline = w_growth_decline.value
        stable_growth = w_stable_growth.value
        stable_growth_year = w_stable_growth_year.value
        ebit_margin = w_ebit_margin.value
        netinv_pct_delta_sales = w_netinv_pct.value
        nwc_pct_delta_sales = w_nwc_pct.value
        tax_rate = w_tax_rate.value
        rwacc = w_rwacc.value
        cash = w_cash.value
        debt = w_debt.value
        shares_out = w_shares_out.value

        # --- Perform Base Calculation ---
        # Unpack the newly added return values
        df_display, terminal_value, enterprise_value, stock_price, error_msg, fcf_terminal_year, pv_fcf, pv_terminal, equity_value_0 = calculate_fcf_valuation(
            sales_2024, start_growth, growth_decline, stable_growth,
            stable_growth_year, ebit_margin, netinv_pct_delta_sales,
            nwc_pct_delta_sales, tax_rate, rwacc,
            cash, debt, shares_out
        )

        # --- Display Results ---
        print("--- Base Case: Multi-Stage FCFF Valuation ---")

        if error_msg:
            print(f"\nCalculation Error: {error_msg}\n")
            # Attempt to display any partial results if available (like df_display)
            if not df_display.empty:
                 print("\n1. Free Cash Flow Forecast (Partial or Complete) ($ Millions)")
                 display(HTML(df_display.to_html(classes='table table-striped table-hover', formatters=format_dict))) # Reuse format dict if possible

        else:
            # Format DataFrame for nice HTML display
            df_display_formatted = df_display.copy()
            format_dict = { # Define format dict here for reuse
                'Growth Rate': '{:.1%}',
                'Sales': '{:,.2f}',
                'Delta Sales': '{:,.2f}',
                'EBIT': '{:,.2f}',
                'Tax': '{:,.2f}',
                'Unlevered NI': '{:,.2f}',
                'Net Investment': '{:,.2f}',
                'Delta NWC': '{:,.2f}',
                'FCF': '{:,.2f}'
            }
            for col, fmt in format_dict.items():
                 if col in df_display_formatted.columns:
                      # Apply formatting safely, handling potential NaNs from errors
                      df_display_formatted[col] = df_display_formatted[col].map(lambda x: fmt.format(x) if pd.notnull(x) else 'N/A')

            print("\n1. Free Cash Flow Forecast ($ Millions)")
            display(HTML(df_display_formatted.to_html(classes='table table-striped table-hover')))

            print(f"\n2. Terminal Value (V_{stable_growth_year-1}) Calculation")
            # Use the returned fcf_terminal_year safely
            print(f"   FCF in {stable_growth_year}: ${fcf_terminal_year:,.2f} M (Using {stable_growth:.1%} growth)")
            print(f"   Terminal Value at end of {stable_growth_year-1}: ${terminal_value:,.2f} M")

            print(f"\n3. Enterprise Value (V_0) Calculation")
            # Use the returned pv_fcf and pv_terminal safely
            print(f"   Present Value of Explicit FCFs (Years {min(df_display.index)}-{max(df_display.index)}): ${pv_fcf:,.2f} M")
            print(f"   Present Value of Terminal Value: ${pv_terminal:,.2f} M")
            print(f"   Estimated Enterprise Value (V₀): ${enterprise_value:,.2f} M")

            print(f"\n4. Stock Price Estimation")
            # Use the returned equity_value_0 safely
            print(f"   Equity Value (V₀ + Cash - Debt): ${equity_value_0:,.2f} M")
            print(f"   Estimated Stock Price (P₀): ${stock_price:,.2f}")

        # --- Calculate and Display Alternative Scenarios ---
        # These calculations remain independent
        print("\n--- Alternative Scenario 1: Constant Growth ---")
        alt1_V0, alt1_P0, alt1_err = calculate_constant_growth_scenario(
             sales_2024, stable_growth, ebit_margin, netinv_pct_delta_sales,
             nwc_pct_delta_sales, tax_rate, rwacc, cash, debt, shares_out)
        if alt1_err:
            print(f"   Calculation Error: {alt1_err}")
        elif pd.notnull(alt1_P0): # Check if result is valid before printing
            print(f"   Enterprise Value (V₀): ${alt1_V0:,.2f} M")
            print(f"   Estimated Stock Price (P₀): ${alt1_P0:,.2f}")
        else:
             print("   Could not calculate Scenario 1 result.")


        print("\n--- Alternative Scenario 2: Constant Growth & Lower Margin (8%) ---")
        low_margin = 0.08 # Hardcode for this scenario
        alt2_V0, alt2_P0, alt2_err = calculate_constant_growth_low_margin_scenario(
             sales_2024, stable_growth, low_margin, netinv_pct_delta_sales,
             nwc_pct_delta_sales, tax_rate, rwacc, cash, debt, shares_out)
        if alt2_err:
            print(f"   Calculation Error: {alt2_err}")
        elif pd.notnull(alt2_P0): # Check if result is valid before printing
            print(f"   Enterprise Value (V₀): ${alt2_V0:,.2f} M")
            print(f"   Estimated Stock Price (P₀): ${alt2_P0:,.2f}")
        else:
            print("   Could not calculate Scenario 2 result.")


# ─── UI Layout ──────────────────────────────────────────────────────────────
recalc_button.on_click(display_valuation)

# Organize widgets into logical groups
input_growth = widgets.VBox([w_sales_2024, w_start_growth, w_growth_decline, w_stable_growth, w_stable_growth_year])
input_ops = widgets.VBox([w_ebit_margin, w_netinv_pct, w_nwc_pct, w_tax_rate])
input_valuation = widgets.VBox([w_rwacc, w_cash, w_debt, w_shares_out])

# Use HBox for side-by-side layout
ui = widgets.VBox([
    widgets.HTML("<h2>AGE Corp - FCFF Valuation Inputs</h2>"),
    widgets.HBox([input_growth, input_ops, input_valuation]),
    recalc_button,
    widgets.HTML("<hr><h3>Valuation Results</h3>"), # Separator
    output_area
])

# ─── Initial Display ────────────────────────────────────────────────────────
display(ui)
display_valuation(None) # Run calculation once on load
````

````python
# If you don’t already have ipywidgets installed, uncomment and run:
# !pip install ipywidgets

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, clear_output, HTML

# Set default style for matplotlib plots
plt.style.use('seaborn-v0_8-darkgrid') # Use a nice style

# ─── Core Calculation Function (Removed RONIC from return) ──────────────────
def calculate_fcf_valuation(
    sales_2024, start_growth, growth_decline, stable_growth,
    stable_growth_year, ebit_margin, netinv_pct_delta_sales,
    nwc_pct_delta_sales, tax_rate, rwacc,
    cash, debt, shares_out):
    """
    Performs the multi-stage FCFF valuation based on inputs.
    Returns: DataFrame, Terminal Value, Enterprise Value, Stock Price, Error Msg,
             FCF_TerminalYear, PV_FCFs, PV_Terminal, Equity_Value
    """
    # --- Setup Years and Growth Rates ---
    forecast_start_year = 2025
    last_explicit_year = stable_growth_year - 1
    if stable_growth_year <= forecast_start_year:
         return pd.DataFrame(), np.nan, np.nan, np.nan, "Error: Stable growth year must be after forecast start year (2025).", np.nan, np.nan, np.nan, np.nan
    forecast_years = list(range(forecast_start_year, stable_growth_year))
    first_stable_year = stable_growth_year
    growth_rates = {}
    current_growth = start_growth
    for year in forecast_years:
        growth_rates[year] = current_growth
        current_growth -= growth_decline
        current_growth = max(current_growth, stable_growth)
    growth_rates[first_stable_year] = stable_growth

    # --- Project Financials ---
    forecast_data = {}
    last_sales = sales_2024
    all_calc_years = list(range(forecast_start_year, first_stable_year + 1))
    for year in all_calc_years:
        data = {}
        data['Growth Rate'] = growth_rates[year]
        if last_sales < 0 or ebit_margin < -1 or tax_rate < 0 or tax_rate > 1:
             return pd.DataFrame(), np.nan, np.nan, np.nan, f"Error: Invalid input detected for year {year} calculation.", np.nan, np.nan, np.nan, np.nan
        data['Sales'] = last_sales * (1 + data['Growth Rate'])
        data['Delta Sales'] = data['Sales'] - last_sales
        data['EBIT'] = data['Sales'] * ebit_margin
        data['Tax'] = data['EBIT'] * tax_rate
        data['Unlevered NI'] = data['EBIT'] * (1 - tax_rate)
        delta_sales_for_inv = max(0, data['Delta Sales'])
        data['Net Investment'] = delta_sales_for_inv * netinv_pct_delta_sales
        data['Delta NWC'] = delta_sales_for_inv * nwc_pct_delta_sales
        data['FCF'] = data['Unlevered NI'] - data['Net Investment'] - data['Delta NWC']
        forecast_data[year] = data
        last_sales = data['Sales']

    df_forecast = pd.DataFrame(forecast_data).T
    valid_forecast_years = [yr for yr in forecast_years if yr in df_forecast.index]
    df_display = df_forecast.loc[valid_forecast_years].copy() if valid_forecast_years else pd.DataFrame()

    if first_stable_year not in forecast_data:
         error_message = f"Error: Data for stable growth year {first_stable_year} could not be calculated."
         return df_display, np.nan, np.nan, np.nan, error_message, np.nan, np.nan, np.nan, np.nan

    fcf_terminal_year = forecast_data[first_stable_year]['FCF']
    g = stable_growth

    # --- Terminal Value Calculation ---
    if rwacc <= g:
        terminal_value_N = np.nan; enterprise_value_0 = np.nan; stock_price_0 = np.nan
        error_message = f"Error: WACC ({rwacc:.2%}) must be greater than Stable Growth Rate ({g:.2%})"
        return df_display, terminal_value_N, enterprise_value_0, stock_price_0, error_message, fcf_terminal_year, np.nan, np.nan, np.nan
    elif np.isclose(rwacc, g):
        terminal_value_N = np.nan; enterprise_value_0 = np.nan; stock_price_0 = np.nan
        error_message = f"Error: WACC ({rwacc:.2%}) is extremely close to Stable Growth Rate ({g:.2%}), leading to unstable Terminal Value."
        return df_display, terminal_value_N, enterprise_value_0, stock_price_0, error_message, fcf_terminal_year, np.nan, np.nan, np.nan
    else:
        error_message = None
        terminal_value_N = fcf_terminal_year / (rwacc - g)

        # --- Enterprise Value Calculation ---
        pv_fcf = 0
        if rwacc <= -1:
            error_message = f"Error: WACC ({rwacc:.2%}) must be greater than -100% for discounting."
            return df_display, terminal_value_N, np.nan, np.nan, error_message, fcf_terminal_year, np.nan, np.nan, np.nan

        for i, year in enumerate(valid_forecast_years):
             if year not in forecast_data or 'FCF' not in forecast_data[year] or pd.isnull(forecast_data[year]['FCF']):
                 error_message = f"Error: Missing or invalid FCF data for year {year}."
                 return df_display, terminal_value_N, np.nan, np.nan, error_message, fcf_terminal_year, pv_fcf if i > 0 else np.nan, np.nan, np.nan
             pv_fcf += forecast_data[year]['FCF'] / ((1 + rwacc)**(i + 1))

        n_explicit = len(valid_forecast_years)
        pv_terminal = terminal_value_N / ((1 + rwacc)**n_explicit)
        enterprise_value_0 = pv_fcf + pv_terminal

        # --- Stock Price ---
        equity_value_0 = enterprise_value_0 + cash - debt
        if shares_out <= 0:
             stock_price_0 = np.nan
             error_message = "Error: Shares Outstanding must be positive." if error_message is None else error_message
        else:
            stock_price_0 = equity_value_0 / shares_out

        # Return values (no implied_ronic)
        return df_display, terminal_value_N, enterprise_value_0, stock_price_0, error_message, fcf_terminal_year, pv_fcf, pv_terminal, equity_value_0


# ─── Sensitivity Plotting Function (Tax Rate instead of NetInv)───────────────
def plot_sensitivity_analysis(base_inputs, output_widget):
    """ Generates and displays the 2x2 sensitivity plots """
    with output_widget:
        n_points = 31

        # Ranges for WACC, Stable Growth, EBIT Margin (as before)
        wacc_lower_bound = base_inputs['stable_growth'] + 0.01
        wacc_plot_min = max(wacc_lower_bound, base_inputs['rwacc'] - 0.04)
        wacc_plot_max = base_inputs['rwacc'] + 0.04
        if wacc_plot_max <= wacc_plot_min: wacc_plot_max = wacc_plot_min + 0.01
        rwacc_range = np.linspace(wacc_plot_min, wacc_plot_max, n_points)

        g_max_bound = base_inputs['rwacc'] - 0.01
        g_plot_min = 0.00
        if g_max_bound <= g_plot_min: g_max_bound = g_plot_min + 0.005
        g_range = np.linspace(g_plot_min, g_max_bound, n_points)

        margin_range = np.linspace(max(0.01, base_inputs['ebit_margin'] * 0.5), base_inputs['ebit_margin'] * 1.5, n_points)

        # --- NEW: Tax Rate Range ---
        tax_plot_min = 0.10 # Example: 10% minimum tax rate
        tax_plot_max = 0.50 # Example: 50% maximum tax rate
        tax_range = np.linspace(tax_plot_min, tax_plot_max, n_points)

        # Store results (replace netinv with tax)
        p0_vs_rwacc, p0_vs_g, p0_vs_margin, p0_vs_tax = [], [], [], []

        # --- Run Sensitivity Calculations ---
        # Unpack one less value now
        for r_test in rwacc_range: _,_,_,p0,err,_,_,_,_ = calculate_fcf_valuation(**{**base_inputs, 'rwacc': r_test}); p0_vs_rwacc.append(p0)
        for g_test in g_range: _,_,_,p0,err,_,_,_,_ = calculate_fcf_valuation(**{**base_inputs, 'stable_growth': g_test}); p0_vs_g.append(p0)
        for m_test in margin_range: _,_,_,p0,err,_,_,_,_ = calculate_fcf_valuation(**{**base_inputs, 'ebit_margin': m_test}); p0_vs_margin.append(p0)
        # NEW: Calculate sensitivity to Tax Rate
        for t_test in tax_range: _,_,_,p0,err,_,_,_,_ = calculate_fcf_valuation(**{**base_inputs, 'tax_rate': t_test}); p0_vs_tax.append(p0)


        p0_vs_rwacc=np.array(p0_vs_rwacc); p0_vs_g=np.array(p0_vs_g); p0_vs_margin=np.array(p0_vs_margin); p0_vs_tax=np.array(p0_vs_tax) # Added tax results
        fig, axs = plt.subplots(2, 2, figsize=(11, 9)); fig.suptitle('Stock Price Sensitivity Analysis', fontsize=16)

        # Plot vs WACC (axs[0, 0] - unchanged)
        valid_rwacc=~np.isnan(p0_vs_rwacc)
        if np.any(valid_rwacc): axs[0,0].plot(rwacc_range[valid_rwacc]*100, p0_vs_rwacc[valid_rwacc], marker='.', markersize=4, linestyle='-', color='crimson')
        axs[0,0].axvline(base_inputs['rwacc']*100, color='grey', linestyle='--', label=f"Base: {base_inputs['rwacc']:.2%}"); axs[0,0].set_xlabel('WACC (%)'); axs[0,0].set_ylabel('Estimated Stock Price ($)'); axs[0,0].set_title('Sensitivity to WACC ($r_{wacc}$)'); axs[0,0].legend(); axs[0,0].tick_params(axis='x', rotation=30); axs[0,0].set_ylim(bottom=0)

        # Plot vs Stable Growth (g) (axs[0, 1] - unchanged)
        valid_g=~np.isnan(p0_vs_g);
        if np.any(valid_g): axs[0,1].plot(g_range[valid_g]*100, p0_vs_g[valid_g], marker='.', markersize=4, linestyle='-', color='forestgreen'); valid_p0_g=p0_vs_g[valid_g]; max_y_g=np.nanpercentile(valid_p0_g, 98)*1.2; min_y_g=np.nanmin(valid_p0_g); axs[0,1].set_ylim(bottom=max(0, min_y_g - 5), top=max_y_g if max_y_g>(min_y_g-5) else min_y_g+10)
        axs[0,1].axvline(base_inputs['stable_growth']*100, color='grey', linestyle='--', label=f"Base: {base_inputs['stable_growth']:.2%}"); axs[0,1].set_xlabel('Stable Growth Rate, g (%)'); axs[0,1].set_ylabel('Estimated Stock Price ($)'); axs[0,1].set_title('Sensitivity to Stable Growth (g)'); axs[0,1].legend(); axs[0,1].tick_params(axis='x', rotation=30)

        # Plot vs EBIT Margin (axs[1, 0] - unchanged)
        valid_margin=~np.isnan(p0_vs_margin);
        if np.any(valid_margin): axs[1,0].plot(margin_range[valid_margin]*100, p0_vs_margin[valid_margin], marker='.', markersize=4, linestyle='-', color='darkorange')
        axs[1,0].axvline(base_inputs['ebit_margin']*100, color='grey', linestyle='--', label=f"Base: {base_inputs['ebit_margin']:.1%}"); axs[1,0].set_xlabel('EBIT Margin (%)'); axs[1,0].set_ylabel('Estimated Stock Price ($)'); axs[1,0].set_title('Sensitivity to EBIT Margin'); axs[1,0].legend(); axs[1,0].tick_params(axis='x', rotation=30); axs[1,0].set_ylim(bottom=0)

        # --- NEW: Plot vs Tax Rate --- (axs[1, 1])
        valid_tax=~np.isnan(p0_vs_tax)
        if np.any(valid_tax):
             axs[1,1].plot(tax_range[valid_tax]*100, p0_vs_tax[valid_tax], marker='.', markersize=4, linestyle='-', color='purple') # New color
        axs[1,1].axvline(base_inputs['tax_rate']*100, color='grey', linestyle='--', label=f"Base: {base_inputs['tax_rate']:.1%}")
        axs[1,1].set_xlabel('Tax Rate, τ (%)'); axs[1,1].set_ylabel('Estimated Stock Price ($)')
        axs[1,1].set_title('Sensitivity to Tax Rate (τ)'); axs[1,1].legend(); axs[1,1].tick_params(axis='x', rotation=30)
        axs[1,1].set_ylim(bottom=0) # Set bottom limit to 0

        plt.tight_layout(rect=[0, 0.03, 1, 0.96])
        plt.show()


# ─── Widget Definitions ─────────────────────────────────────────────────────
# (Widgets remain the same)
style = {'description_width': 'initial'}
layout = widgets.Layout(width='350px')
w_sales_2024 = widgets.FloatText(value=600.0, description='Sales 2024 ($M):', style=style, layout=layout)
w_start_growth = widgets.FloatText(value=0.08, step=0.005, description='Growth Rate 2025:', style=style, layout=layout)
w_growth_decline = widgets.FloatText(value=0.01, step=0.001, description='Annual Growth Decline:', style=style, layout=layout)
w_stable_growth = widgets.FloatText(value=0.035, step=0.001, description='Stable Growth Rate (g):', style=style, layout=layout)
w_stable_growth_year = widgets.IntText(value=2030, step=1, description='Year Stable Growth Begins:', style=style, layout=layout)
w_ebit_margin = widgets.FloatText(value=0.10, step=0.005, description='EBIT Margin (% of Sales):', style=style, layout=layout)
w_netinv_pct = widgets.FloatText(value=0.07, step=0.005, description='Net Inv (% of ΔSales):', style=style, layout=layout)
w_nwc_pct = widgets.FloatText(value=0.12, step=0.005, description='ΔNWC (% of ΔSales):', style=style, layout=layout)
w_tax_rate = widgets.FloatText(value=0.35, step=0.01, description='Tax Rate (τ):', style=style, layout=layout)
w_rwacc = widgets.FloatText(value=0.10, step=0.001, description='WACC (rwacc):', style=style, layout=layout)
w_cash = widgets.FloatText(value=80.0, description='Cash ($M):', style=style, layout=layout)
w_debt = widgets.FloatText(value=10.0, description='Debt ($M):', style=style, layout=layout)
w_shares_out = widgets.FloatText(value=30.0, step=0.1, description='Shares Out (Millions):', style=style, layout=layout)
recalc_button = widgets.Button(description='Calculate Valuation', button_style='info', layout=widgets.Layout(width='200px'))
output_area = widgets.Output()
sensitivity_output_area = widgets.Output()


# ─── Display Logic (Removed RONIC check)─────────────────────────────────
def display_valuation(_): # Triggered by button
    with output_area:
        clear_output(wait=True)
        base_inputs = {
            'sales_2024': w_sales_2024.value, 'start_growth': w_start_growth.value,
            'growth_decline': w_growth_decline.value, 'stable_growth': w_stable_growth.value,
            'stable_growth_year': w_stable_growth_year.value, 'ebit_margin': w_ebit_margin.value,
            'netinv_pct_delta_sales': w_netinv_pct.value, 'nwc_pct_delta_sales': w_nwc_pct.value,
            'tax_rate': w_tax_rate.value, 'rwacc': w_rwacc.value, 'cash': w_cash.value,
            'debt': w_debt.value, 'shares_out': w_shares_out.value
        }

        # --- Perform Base Calculation ---
        # Unpack results (no implied_ronic now)
        df_display, terminal_value, enterprise_value, stock_price, error_msg, \
        fcf_terminal_year, pv_fcf, pv_terminal, equity_value_0 = calculate_fcf_valuation(**base_inputs)

        # --- Display Text Results ---
        print("--- Base Case: Multi-Stage FCFF Valuation ---")
        if error_msg:
            print(f"\nCalculation Error: {error_msg}\n")
            with sensitivity_output_area: clear_output()
        else:
            # Format DataFrame
            df_display_formatted = df_display.copy()
            format_dict = { 'Growth Rate': '{:.1%}', 'Sales': '{:,.2f}', 'Delta Sales': '{:,.2f}', 'EBIT': '{:,.2f}', 'Tax': '{:,.2f}', 'Unlevered NI': '{:,.2f}', 'Net Investment': '{:,.2f}', 'Delta NWC': '{:,.2f}', 'FCF': '{:,.2f}'}
            if not df_display.empty:
                for col, fmt in format_dict.items():
                    if col in df_display_formatted.columns: df_display_formatted[col] = df_display_formatted[col].map(lambda x: fmt.format(x) if pd.notnull(x) else 'N/A')
                print("\n1. Free Cash Flow Forecast ($ Millions)")
                display(HTML(df_display_formatted.to_html(classes='table table-striped table-hover')))
            else: print("\n1. Free Cash Flow Forecast could not be generated.")

            # Display steps 2 & 3
            if pd.notnull(terminal_value):
                 print(f"\n2. Terminal Value (V_{base_inputs['stable_growth_year']-1}) Calculation")
                 print(f"   FCF in {base_inputs['stable_growth_year']}: ${fcf_terminal_year:,.2f} M (Using {base_inputs['stable_growth']:.1%} growth)")
                 print(f"   Terminal Value at end of {base_inputs['stable_growth_year']-1}: ${terminal_value:,.2f} M")
            if pd.notnull(enterprise_value):
                 print(f"\n3. Enterprise Value (V_0) Calculation")
                 print(f"   Present Value of Explicit FCFs (Years {min(df_display.index) if not df_display.empty else 'N/A'}-{max(df_display.index) if not df_display.empty else 'N/A'}): ${pv_fcf:,.2f} M")
                 print(f"   Present Value of Terminal Value: ${pv_terminal:,.2f} M")
                 print(f"   Estimated Enterprise Value (V₀): ${enterprise_value:,.2f} M")

            # --- REMOVED Stable Phase Checks / RONIC display ---

            # Display Stock Price
            if pd.notnull(stock_price):
                 print(f"\n4. Stock Price Estimation")
                 print(f"   Equity Value (V₀ + Cash - Debt): ${equity_value_0:,.2f} M")
                 print(f"   Estimated Stock Price (P₀): ${stock_price:,.2f}")

            # --- Call Sensitivity Plot Function ---
            with sensitivity_output_area:
                 clear_output(wait=True)
                 plot_sensitivity_analysis(base_inputs, sensitivity_output_area)


# ─── UI Layout ──────────────────────────────────────────────────────────────
recalc_button.on_click(display_valuation)
input_growth = widgets.VBox([w_sales_2024, w_start_growth, w_growth_decline, w_stable_growth, w_stable_growth_year])
input_ops = widgets.VBox([w_ebit_margin, w_netinv_pct, w_nwc_pct, w_tax_rate])
input_valuation = widgets.VBox([w_rwacc, w_cash, w_debt, w_shares_out])
ui = widgets.VBox([
    widgets.HTML("<h2>AGE Corp - FCFF Valuation Inputs</h2>"),
    widgets.HBox([input_growth, input_ops, input_valuation]),
    recalc_button,
    widgets.HTML("<hr><h3>Valuation Results</h3>"),
    output_area, # Area for text results
    widgets.HTML("<hr><h3>Sensitivity Analysis</h3>"),
    sensitivity_output_area # Area for plots
])

# ─── Initial Display ────────────────────────────────────────────────────────
display(ui)
display_valuation(None) # Run calculation and plotting once on load
````

### Valuation Based on Comparable Firms
*   We estimate the value of the firm based on the value of other, comparable firms or investments that we expect will generate very similar cash flows in the future rather than value the firm's cash flows directly.
    *   Challenge is finding comparable firms or investments.
    *   The differences in scale between firms can be adjusted by expressing their value in terms of a valuation multiple.
    *   A valuation multiple is a ratio of the value to some measure of the firm's scale.

**Valuation Multiples**
*   The Price-Earnings Ratio (P/E Ratio)
*   Enterprise Value Multiples
*   Other Multiples
    *   Multiple of sales
    *   Price to book value of equity per share
    *   Enterprise value per subscriber

---

### Valuation Based on Comparable Firms: The Price-Earnings Ratio (P/E Ratio)
*   Share price divided by earnings per share (EPS)
    *   When you buy a stock, you are in a sense buying the rights to the firm's future earnings.
*   Trailing P/E: Uses earnings over the last 12 months
*   Forward P/E: Uses expected earnings over the next 12 months

**Relationship to Dividend-Discount Model:**

$$ \text{Forward } P/E = \frac{P_0}{EPS_1} = \frac{Div_1 / EPS_1}{r_E - g} = \frac{\text{Dividend Payout Ratio}}{r_E - g} $$

*   If two stocks have the same payout and EPS growth rates, as well as equivalent risk (implying same $r_E$), then they should have the same P/E.
*   Firms and industries with high growth rates, and that generate cash well in excess of their investment needs so that they can maintain high payout ratios, should have high P/E multiples.

---

### Valuation Based on Comparable Firms: The Price-Earnings Ratio (P/E Ratio)

* Suppose industrial equipment manufacturer BSM Corp has earnings per share of $\$2.55$. If the average price-to-earnings (P/E) ratio of comparable industrial equipment stocks is $18.2$, estimate a value for BSM Corp using the P/E as a valuation multiple. What are the assumptions underlying this estimate?

    * $\text{Given:} \quad P/E_{comp} = 18.2 \quad \text{and} \quad EPS_{BSM} = \$2.55$.
    * Estimate $P_{BSM}$ (Price of BSM Corp stock):
        The relationship between price, EPS, and P/E ratio is:
        $$ \frac{P_{BSM}}{EPS_{BSM}} \approx P/E_{comp} $$
                
        $ P_{BSM} \approx P/E_{comp} \times EPS_{BSM} \qquad \Rightarrow \qquad P_{BSM} \approx 18.2 \times \$2.55 = \$46.41 $
    
    * Assumption: BSM Corp is, on average, similar to the comparable firms used to calculate the average P/E ratio. 
        * Specifically, it assumes BSM Corp shares key value drivers with these firms, including:
            * Similar risk (implying a similar equity cost of capital, $r_E$).
            * Similar future growth prospects ($g$).
            * Similar accounting methods.
            * Similar payout rates or reinvestment policies.

### Valuation Based on Comparable Firms: The Price-Earnings Ratio (P/E Ratio)

You observe that AGE Corp (AGE) has a stock price of $\$95.50$ and reports EPS of $\$4.20$. Its competitor, PSK Ltd (PSK), has an EPS of $\$3.15$. Estimate the value of a share of PSK stock using only this data and the method of comparables.

$$ (P/E)_{AGE} = \frac{P_{AGE}}{EPS_{AGE}} = \frac{\$95.50}{\$4.20} \approx 22.74 $$

$$ P_{PSK} = (P/E)_{AGE} \times EPS_{PSK} $$

$$ P_{PSK} \approx 22.74 \times \$3.15 \approx \$71.63 $$

### Valuation Based on Comparable Firms: Enterprise Value Multiples

*   Valuation multiples based on the firm's enterprise value.

*   Using the enterprise value is advantageous if we want to compare firms with different amounts of leverage.

*   Since the enterprise value represents the entire value of the firm before the firm pays its debt, to form an appropriate multiple, we divide it by a measure of earnings or cash flows before interest payments are made (i.e. EBIT, EBITDA).

*   If expected free cash flow growth is constant,
    
    $$ \frac{V_0}{EBITDA_1} = \frac{FCF_1 / EBITDA_1}{r_{wacc} - g_{FCF}} $$
    
    where $V_0$ is Enterprise Value.

### Valuation Based on Comparable Firms: Enterprise Value Multiples

Suppose Alpha Corp has earnings per share (EPS) of $\$3.50$ and EBITDA of $\$45.0$ million. Alpha Corp also has $8.0$ million shares outstanding and net debt of $\$150$ million. You believe Beta Corp is comparable to Alpha Corp in terms of its underlying business, but Beta Corp has very little debt compared to Alpha Corp.

If Beta Corp trades at a P/E ratio of $15.0$ and an EV/EBITDA multiple of $8.5$, estimate the value of Alpha Corp's shares using both multiples. Discuss which estimate is likely to be more reliable.

$P/E_{Beta} = 15.0 \quad \text{and} \quad EPS_{Alpha} = \$3.50 \qquad \Rightarrow \qquad P^{Alpha}_0 = P/E_{Beta} \times EPS_{Alpha} = 15.0 \times \$3.50 = \$52.50 $

**EV/EBITDA Multiple Valuation**

$(EV/EBITDA)_{Beta} = 8.5 \quad \text{and} \quad EBITDA_{Alpha} = \$45.0 \text{ million}$

$\frac{EV_{Alpha}}{EBITDA_{Alpha}} \approx (EV/EBITDA)_{Beta} \qquad \Rightarrow \qquad EV_{Alpha} \approx 8.5 \times EBITDA_{Alpha} = 8.5 \times \$45.0 \text{ million} = \$382.5 \text{ million} $

$ \text{Equity Value}_{Alpha} = EV_{Alpha} - \text{Net Debt}_{Alpha} \qquad \Rightarrow \qquad \text{Equity Value}_{Alpha} = \$382.5 \text{ million} - \$150 \text{ million} = \$232.5 \text{ million} $

$ P^{Alpha}_0 = \frac{\text{Equity Value}_{Alpha}}{\text{Shares Outstanding}_{Alpha}} = \frac{\$232.5 \text{ million}}{8.0 \text{ million}} \approx \$29.06 $

**Comparison of Estimates**

The key difference between the firms is their capital structure: Alpha Corp has significant net debt ($\$150 \, \text{million}$), while Beta Corp has little debt.
* P/E ratios are calculated based on net earnings (after interest expense) and market capitalization (equity value). High debt increases interest expense, reducing net earnings and potentially distorting the P/E ratio when comparing firms with different leverage levels.
* EV/EBITDA multiples use metrics that are independent of capital structure. Enterprise Value includes both debt and equity, and EBITDA is calculated before interest expense. This makes the EV/EBITDA multiple less sensitive to differences in leverage between comparable firms.

### Valuation Based on Comparable Firms: Other Multiples and Limitations of Multiples

*   Multiple of sales
*   Price to book value of equity per share
*   Enterprise value per subscriber

**Limitations**
*   Firms are not identical.
    *   The usefulness of a valuation multiple will depend on the nature of the differences between firms and the sensitivity of the multiples to these differences.
    *   When valuing a firm using multiples, there is no clear guidance about how to adjust for these differences.
*   Comparables only provide information regarding the value of a firm relative to the other firms in the comparison set.
*   Each multiple has a different set of value drivers.
    $$ \text{Forward } P/E = \frac{P_0}{EPS_1} = \frac{\text{Dividend Payout Rate}}{r_E - g} $$

**Stock Valuation Techniques**
*   No single technique provides a final answer regarding a stock's true (intrinsic) value.
    *   All approaches require assumptions or forecasts that are too uncertain to provide a definitive assessment of the firm's value.
    *   In practice, a combination of these methods is used.
