**Question 1:** 1. If the first-year interest rate is 2% and the second year interest is 3%, what is the 2-year total interest rate?

**Answer:** The 2-year total interest rate is 5.06%.

To calculate the total interest rate over two years with varying annual interest rates, we must consider the effect of compound interest.  Let $r_1$ be the interest rate in the first year and $r_2$ be the interest rate in the second year.  Assume an initial principal amount $P$.

In the first year, the interest earned is calculated as:
$$I_1 = P \times r_1$$
The amount at the end of the first year, $A_1$, is the principal plus the interest earned:
$$A_1 = P + I_1 = P + P \times r_1 = P(1 + r_1)$$

In the second year, the interest is calculated on the amount accumulated at the end of the first year, $A_1$. The interest earned in the second year is:
$$I_2 = A_1 \times r_2 = P(1 + r_1) \times r_2$$
The amount at the end of the second year, $A_2$, is the amount at the end of the first year plus the interest earned in the second year:
$$A_2 = A_1 + I_2 = P(1 + r_1) + P(1 + r_1) \times r_2 = P(1 + r_1)(1 + r_2)$$

The total interest earned over two years is the difference between the final amount and the initial principal:
$$I_{total} = A_2 - P = P(1 + r_1)(1 + r_2) - P = P[(1 + r_1)(1 + r_2) - 1]$$

The total interest rate over two years is the total interest earned divided by the initial principal, expressed as a percentage:
$$R_{total} = \frac{I_{total}}{P} \times 100\% = \frac{P[(1 + r_1)(1 + r_2) - 1]}{P} \times 100\% = [(1 + r_1)(1 + r_2) - 1] \times 100\%$$

Given $r_1 = 2\% = 0.02$ and $r_2 = 3\% = 0.03$, we can substitute these values into the formula:
$$R_{total} = [(1 + 0.02)(1 + 0.03) - 1] \times 100\%$$
$$R_{total} = [(1.02)(1.03) - 1] \times 100\%$$
$$R_{total} = [1.0506 - 1] \times 100\%$$
$$R_{total} = 0.0506 \times 100\%$$
$$R_{total} = 5.06\%$$

Therefore, the 2-year total interest rate is 5.06%.  It is important to note that this is not simply the sum of the individual yearly interest rates (2% + 3% = 5%) due to the effect of compounding, where interest earned in the first year also earns interest in the second year.

```

**Question 5:** 5. If you earn a rate of return of 5% over 4 months, what is the annualized rate of return?

**Answer:** The annualized rate of return can be calculated by compounding the periodic rate of return over the number of periods in a year.
Given a rate of return of 5% over a period of 4 months, we first identify the periodic rate and the length of the period.

The periodic rate of return ($r_p$) is 5%, which is 0.05 in decimal form.
The period is 4 months.

To annualize this rate, we need to determine how many such periods are there in a year.
Since there are 12 months in a year, the number of 4-month periods in a year ($n$) is:
$$n = \frac{12 \text{ months}}{4 \text{ months}} = 3$$

Assuming the return is compounded over these periods, the annualized rate of return ($r_a$) can be found using the formula:
$$1 + r_a = (1 + r_p)^n$$

Substituting the values of $r_p$ and $n$ into the formula:
$$1 + r_a = (1 + 0.05)^3$$
$$1 + r_a = (1.05)^3$$

Now we calculate $(1.05)^3$:
$$(1.05)^3 = 1.05 \times 1.05 \times 1.05$$
$$(1.05)^2 = 1.1025$$
$$(1.05)^3 = 1.1025 \times 1.05 = 1.157625$$

So,
$$1 + r_a = 1.157625$$

To find $r_a$, we subtract 1 from both sides:
$$r_a = 1.157625 - 1$$
$$r_a = 0.157625$$

To express the annualized rate of return as a percentage, we multiply $r_a$ by 100:
$$r_a \% = 0.157625 \times 100\% = 15.7625\%$$

Therefore, the annualized rate of return is 15.7625%.  For practical purposes, this may be rounded to two decimal places, giving 15.76%.

Final Answer: The final answer is $\boxed{15.7625\%}$

**Question 6:** 6. Assume that the 2-year holding rate of return is 40%. The average (arithmetic) rate of return is therefore 20% per year. What is the annualized (geometric) rate of return? Is the annualized rate the same as the average rate?

**Answer:** The annualized (geometric) rate of return is calculated as follows:
Given a 2-year holding rate of return ($HPR$) of 40% or 0.40, we can represent the relationship between the initial investment and the final value after 2 years as:
$Final\ Value = Initial\ Investment \times (1 + HPR)$

Let $r_g$ be the annualized (geometric) rate of return.  For a 2-year period, the final value can also be expressed as:
$Final\ Value = Initial\ Investment \times (1 + r_g) \times (1 + r_g) = Initial\ Investment \times (1 + r_g)^2$

Equating the two expressions for the final value, we have:
$Initial\ Investment \times (1 + HPR) = Initial\ Investment \times (1 + r_g)^2$

Dividing both sides by the Initial Investment, we get:
$1 + HPR = (1 + r_g)^2$

Substituting the given $HPR = 0.40$:
$1 + 0.40 = (1 + r_g)^2$
$1.40 = (1 + r_g)^2$

To solve for $r_g$, we take the square root of both sides:
$\sqrt{1.40} = 1 + r_g$

Subtracting 1 from both sides to isolate $r_g$:
$r_g = \sqrt{1.40} - 1$

Calculating the value of $\sqrt{1.40}$:
$\sqrt{1.40} \approx 1.1832$

Therefore, the annualized (geometric) rate of return is:
$r_g \approx 1.1832 - 1 = 0.1832$

Converting this to percentage form, we multiply by 100:
$r_g \approx 0.1832 \times 100 = 18.32\%$

The annualized (geometric) rate of return is approximately 18.32%.

The average (arithmetic) rate of return is given as 20% per year.
Comparing the annualized (geometric) rate of return (18.32%) with the average (arithmetic) rate of return (20%), we can see that they are not the same. The annualized (geometric) rate of return is lower than the average (arithmetic) rate of return in this case.  This is generally true because the geometric mean accounts for the effect of compounding, and it is always less than or equal to the arithmetic mean, with equality holding only when all returns are the same.

Final answer: The annualized (geometric) rate of return is approximately 18.32%. The annualized rate is not the same as the average rate.

Final Answer: The annualized (geometric) rate of return is approximately 18.32%. The annualized rate is not the same as the average rate.

**Question 7:** 7. Is the compounded rate of return higher or lower than the sum of the individual rates of return? Is the annualized rate of return higher or lower than the average of the individual rates of return? Why?

**Answer:** The compounded rate of return is generally different from the sum of the individual rates of return. For periods with non-zero returns, the compounded rate of return will not be equal to the sum of the individual rates of return, and for typical positive returns, it will often be numerically close to, but not directly comparable to, the sum.

The annualized rate of return is generally lower than or equal to the average of the individual rates of return.

**Explanation:**

**Compounded Rate of Return vs. Sum of Individual Rates of Return**

The compounded rate of return reflects the cumulative effect of returns over multiple periods, where returns in each period are earned not only on the initial investment but also on the accumulated returns from previous periods. This is in contrast to simply summing individual rates of return, which does not account for this compounding effect.

Consider an investment over two periods with rates of return $r_1$ and $r_2$ in period 1 and period 2 respectively. Let $P_0$ be the initial investment.

After period 1, the value of the investment $P_1$ is:
$P_1 = P_0 (1 + r_1)$

After period 2, the value of the investment $P_2$ is:
$P_2 = P_1 (1 + r_2) = P_0 (1 + r_1)(1 + r_2)$

The overall growth factor over the two periods is $\frac{P_2}{P_0} = (1 + r_1)(1 + r_2)$.

The compounded rate of return over the two periods, $R_c$, is such that $1 + R_c$ equals this overall growth factor:
$1 + R_c = (1 + r_1)(1 + r_2)$
$R_c = (1 + r_1)(1 + r_2) - 1$
$R_c = 1 + r_1 + r_2 + r_1 r_2 - 1$
$R_c = r_1 + r_2 + r_1 r_2$

The sum of the individual rates of return, $R_s$, is simply:
$R_s = r_1 + r_2$

Comparing $R_c$ and $R_s$, we see that:
$R_c = R_s + r_1 r_2$

The difference between the compounded rate of return and the sum of individual rates of return is the term $r_1 r_2$.

In general, for $n$ periods with individual rates of return $r_1, r_2, ..., r_n$, the compounded rate of return $R_c$ is given by:
$1 + R_c = \prod_{i=1}^{n} (1 + r_i)$
$R_c = \left( \prod_{i=1}^{n} (1 + r_i) \right) - 1$

The sum of individual rates of return is:
$R_s = \sum_{i=1}^{n} r_i$

The compounded rate of return is not simply higher or lower than the sum of individual rates in all cases. The relationship depends on the values and signs of the individual rates of return. However, if all individual rates of return are positive, then the term $r_1 r_2$ (and higher-order product terms for more periods in the expanded product $\prod_{i=1}^{n} (1 + r_i)$) will be positive, making the compounded rate of return different from and typically numerically larger than the sum of the individual rates of return.  If some rates are negative, the relationship can be more complex, but in most practical contexts, they are not equivalent.

**Annualized Rate of Return vs. Average of Individual Rates of Return**

The annualized rate of return is designed to provide a standardized rate of return expressed on a per-year basis, regardless of the investment period. It is derived from the overall compounded return over a period longer or shorter than one year. The average of the individual rates of return is the arithmetic mean of the returns over multiple periods.

Let's consider $n$ years of investment with annual rates of return $r_1, r_2, ..., r_n$. The overall compounded return over these $n$ years, $R_{n}$, is given by:
$1 + R_{n} = \prod_{i=1}^{n} (1 + r_i)$

The annualized rate of return, $R_a$, is the constant annual rate that, if compounded over $n$ years, would yield the same overall growth. Therefore:
$(1 + R_a)^n = 1 + R_{n} = \prod_{i=1}^{n} (1 + r_i)$
Taking the $n$-th root of both sides:
$1 + R_a = \left( \prod_{i=1}^{n} (1 + r_i) \right)^{1/n}$
$R_a = \left( \prod_{i=1}^{n} (1 + r_i) \right)^{1/n} - 1$

The annualized rate of return is effectively calculated using the geometric mean of the growth factors $(1+r_i)$.

The average of the individual rates of return, $R_{avg}$, is the arithmetic mean of the individual annual returns:
$R_{avg} = \frac{1}{n} \sum_{i=1}^{n} r_i$

By the Arithmetic Mean - Geometric Mean (AM-GM) inequality, for a set of non-negative numbers, the arithmetic mean is always greater than or equal to the geometric mean.  Assuming $(1+r_i) \ge 0$ for all $i$ (which is generally true for investment returns unless losses exceed 100%), we have:
$\frac{\sum_{i=1}^{n} (1 + r_i)}{n} \ge \left( \prod_{i=1}^{n} (1 + r_i) \right)^{1/n}$
$\frac{\sum_{i=1}^{n} 1 + \sum_{i=1}^{n} r_i}{n} \ge \left( \prod_{i=1}^{n} (1 + r_i) \right)^{1/n}$
$\frac{n + \sum_{i=1}^{n} r_i}{n} \ge \left( \prod_{i=1}^{n} (1 + r_i) \right)^{1/n}$
$1 + \frac{\sum_{i=1}^{n} r_i}{n} \ge \left( \prod_{i=1}^{n} (1 + r_i) \right)^{1/n}$
$1 + R_{avg} \ge 1 + R_a$
$R_{avg} \ge R_a$

The average of the individual rates of return is always greater than or equal to the annualized rate of return. Equality holds if and only if all individual rates of return are equal ($r_1 = r_2 = ... = r_n$).

Therefore, the annualized rate of return is generally lower than or equal to the average of the individual rates of return because the annualized rate is based on the geometric mean, while the average rate is the arithmetic mean, and the arithmetic mean is always greater than or equal to the geometric mean for non-negative values.

**Question 8:** 8. If the total holding interest rate is 50% for a 5-year investment, what is the annualized rate of return?

**Answer:** The annualized rate of return is approximately 8.45%.

**Explanation:**

Let $H$ be the total holding interest rate over the investment period, $n$ be the number of years of the investment, and $r$ be the annualized rate of return.

The total holding interest rate represents the total percentage increase in the value of an investment over the entire investment period. If the initial investment is $P_0$, and the final value after $n$ years is $P_n$, then the total holding interest rate $H$ is given by:

$H = \frac{P_n - P_0}{P_0}$

This can be rewritten as:

$\frac{P_n}{P_0} = 1 + H$

Assuming compound interest, the relationship between the final value $P_n$, the initial investment $P_0$, the annualized rate of return $r$, and the number of years $n$ is given by the formula:

$P_n = P_0 (1 + r)^n$

Dividing both sides by $P_0$, we get:

$\frac{P_n}{P_0} = (1 + r)^n$

Combining the two expressions for $\frac{P_n}{P_0}$, we have:

$(1 + r)^n = 1 + H$

Given that the total holding interest rate $H = 50\% = 0.50$ and the investment period $n = 5$ years, we can substitute these values into the equation:

$(1 + r)^5 = 1 + 0.50$

$(1 + r)^5 = 1.50$

To solve for $r$, we take the fifth root of both sides:

$1 + r = (1.50)^{\frac{1}{5}}$

$r = (1.50)^{\frac{1}{5}} - 1$

Now, we calculate the value of $(1.50)^{\frac{1}{5}}$:

$(1.50)^{\frac{1}{5}} \approx 1.08447177$

Subtracting 1 from this value, we get:

$r \approx 1.08447177 - 1$

$r \approx 0.08447177$

To express the annualized rate of return as a percentage, we multiply by 100:

$r \approx 0.08447177 \times 100$

$r \approx 8.447177 \%$

Rounding to two decimal places, we get:

$r \approx 8.45 \%$

Therefore, the annualized rate of return is approximately 8.45%.

**Verification:**

To verify, we can calculate the total return over 5 years with an annualized rate of 8.45%:

$(1 + 0.0845)^5 = (1.0845)^5 \approx 1.5002$

The total holding interest rate would be $1.5002 - 1 = 0.5002$, which is approximately 50%. This confirms our calculation.

**Question 9:** 9. If the per-year interest rate is 10% for each of the next 5 years, what is the annualized 5-year rate of return?

**Answer:** ```answer
Assuming an initial investment of \$1, the value after one year with a 10% per-year interest rate is $1 \times (1 + 0.10) = 1.10$.
After two years, the value becomes $1.10 \times (1 + 0.10) = 1.10^2 = 1.21$.
Following this pattern, after five years, the value will be $1 \times (1 + 0.10)^5 = 1.10^5$.

To calculate $1.10^5$:
$1.10^2 = 1.21$
$1.10^3 = 1.21 \times 1.10 = 1.331$
$1.10^4 = 1.331 \times 1.10 = 1.4641$
$1.10^5 = 1.4641 \times 1.10 = 1.61051$

So, after 5 years, an initial investment of \$1 grows to \$1.61051.
The total return over the 5-year period is the ratio of the final amount to the initial amount, which is $1.61051 / 1 = 1.61051$.
This represents a total growth of $1.61051 - 1 = 0.61051$, or $61.051\%$ over 5 years.

To find the annualized 5-year rate of return, we need to find an annual interest rate, $r_a$, such that compounding it over 5 years yields the same total return.  This can be expressed as:
$$(1 + r_a)^5 = 1.61051$$

To solve for $r_a$, we take the fifth root of both sides:
$$1 + r_a = (1.61051)^{1/5}$$

Calculating the fifth root of 1.61051:
$$(1.61051)^{1/5} = 1.10$$

Therefore,
$$1 + r_a = 1.10$$
$$r_a = 1.10 - 1$$
$$r_a = 0.10$$

The annualized 5-year rate of return is $0.10$, or $10\%$.

In this specific scenario, because the per-year interest rate is constant at 10% for each of the 5 years, the annualized 5-year rate of return is simply equal to the per-year interest rate. This is because the definition of an annualized rate, in this context, seeks a constant annual rate that is equivalent to the given series of annual rates over the specified period. When the annual rates are already constant, that constant rate is itself the annualized rate.

Final Answer: The final answer is $\boxed{10\%}$
```

**Question 10:** 10. A project costs $200 and will provide cash flows of +$100, +$300, and +$500 in consecutive years. The annualized interest rate is 3% per annum over 1 year, 4% per annum over 2 years, and 4.5% per annum over 3 years. What is this project's NPV?

**Answer:** The Net Present Value (NPV) of a project is calculated by discounting all future cash flows back to their present value and subtracting the initial investment. In this case, we have an initial cost and positive cash flows in the following three years, with different discount rates for each year.

To calculate the NPV, we will use the formula for present value for each cash flow and then sum them up, including the initial cost. The present value (PV) of a future cash flow (CF) at year $t$ with a discount rate $r_t$ is given by:

$$PV = \frac{CF}{(1+r_t)^t}$$

In this problem, we have:
\begin{itemize}
    \item Initial Investment (Year 0 Cash Flow): $CF_0 = -\$200$
    \item Year 1 Cash Flow: $CF_1 = +\$100$, Discount rate $r_1 = 3\% = 0.03$
    \item Year 2 Cash Flow: $CF_2 = +\$300$, Discount rate $r_2 = 4\% = 0.04$
    \item Year 3 Cash Flow: $CF_3 = +\$500$, Discount rate $r_3 = 4.5\% = 0.045$
\end{itemize}

We calculate the present value for each cash flow:

For Year 1 Cash Flow:
$$PV_1 = \frac{CF_1}{(1+r_1)^1} = \frac{100}{(1+0.03)^1} = \frac{100}{1.03} \approx 97.087$$

For Year 2 Cash Flow:
$$PV_2 = \frac{CF_2}{(1+r_2)^2} = \frac{300}{(1+0.04)^2} = \frac{300}{(1.04)^2} = \frac{300}{1.0816} \approx 277.365$$

For Year 3 Cash Flow:
$$PV_3 = \frac{CF_3}{(1+r_3)^3} = \frac{500}{(1+0.045)^3} = \frac{500}{(1.045)^3} = \frac{500}{1.141166125} \approx 438.143$$

The present value of the initial investment at Year 0 is simply the initial investment itself, as it is already in present value terms:
$$PV_0 = CF_0 = -200$$

The Net Present Value (NPV) is the sum of the present values of all cash flows:
$$NPV = PV_0 + PV_1 + PV_2 + PV_3$$
$$NPV = -200 + 97.087 + 277.365 + 438.143$$
$$NPV \approx 612.595$$

Rounding to two decimal places for currency, the NPV is approximately $612.60$.

Therefore, the project's NPV is approximately $612.60. Since the NPV is positive, the project is considered to be profitable at the given discount rates.

**Question 12:** 12. Write down the relationship between nominal rates of return ($r_{nominal}$), real rates of return ($r_{real}$), and the inflation rate ($\pi$).

**Answer:** The relationship between nominal rates of return ($r_{nominal}$), real rates of return ($r_{real}$), and the inflation rate ($\pi$) is given by the Fisher equation.

**Approximate Relationship:**

A common approximation for the relationship is:

$r_{nominal} \approx r_{real} + \pi$

This approximation suggests that the nominal rate of return is approximately equal to the sum of the real rate of return and the inflation rate.

**Detailed Explanation of the Approximation:**

The intuition behind this approximation is that investors require compensation for both the real increase in their purchasing power (real rate of return) and the erosion of purchasing power due to inflation.  If an investor wants to achieve a certain real return on their investment, they must earn a nominal return that is sufficient to offset the loss of purchasing power caused by inflation and still provide the desired real gain.

For example, if an investor desires a real rate of return of 3% and the expected inflation rate is 2%, the approximate nominal rate of return they would require is $3\% + 2\% = 5\%$.

**Exact Relationship (Fisher Equation):**

The more precise relationship is given by the Fisher equation:

$1 + r_{nominal} = (1 + r_{real})(1 + \pi)$

This equation can be rearranged to solve for the real rate of return:

$r_{real} = \frac{1 + r_{nominal}}{1 + \pi} - 1$

Or to solve for the nominal rate of return:

$r_{nominal} = (1 + r_{real})(1 + \pi) - 1$

**Detailed Explanation of the Fisher Equation:**

The Fisher equation is derived from the concept of purchasing power.  Consider an investment of \$1.  After one period, the nominal value of the investment will be $\$(1 + r_{nominal})$.  However, due to inflation, the price level has increased by a factor of $(1 + \pi)$.  Therefore, to find the real value of the investment in terms of purchasing power, we need to divide the nominal value by the inflation factor $(1 + \pi)$.

The real value of the investment after one period, in terms of initial purchasing power, is $\frac{1 + r_{nominal}}{1 + \pi}$.  The real rate of return is the percentage increase in this real value, which is given by:

$r_{real} = \frac{\text{Real Value at End} - \text{Initial Real Value}}{\text{Initial Real Value}} = \frac{\frac{1 + r_{nominal}}{1 + \pi} - 1}{1} = \frac{1 + r_{nominal}}{1 + \pi} - 1$

**Relationship between the Approximation and the Fisher Equation:**

The approximate formula is a simplification of the Fisher equation, and it works well when inflation rates and real rates of return are relatively low.  We can see how the approximation arises from the Fisher equation by expanding the Fisher equation and simplifying for small values of $r_{real}$ and $\pi$.

Starting from the Fisher equation:
$1 + r_{nominal} = (1 + r_{real})(1 + \pi) = 1 + r_{real} + \pi + r_{real}\pi$

Subtracting 1 from both sides:
$r_{nominal} = r_{real} + \pi + r_{real}\pi$

The term $r_{real}\pi$ is the product of the real rate and the inflation rate. When both $r_{real}$ and $\pi$ are small (e.g., single-digit percentages), their product $r_{real}\pi$ becomes very small and can be neglected without introducing significant error.  For instance, if $r_{real} = 0.03$ and $\pi = 0.02$, then $r_{real}\pi = 0.03 \times 0.02 = 0.0006 = 0.06\%$, which is small compared to $r_{real} = 3\%$ and $\pi = 2\%$.

Therefore, for low inflation and real interest rates, we can approximate the relationship as:
$r_{nominal} \approx r_{real} + \pi$

**When to Use Each Formula:**

* **Approximate Formula ($r_{nominal} \approx r_{real} + \pi$):** This approximation is generally sufficient for quick estimations and when inflation rates are low (e.g., single-digit percentages). It is widely used in introductory economics and finance discussions due to its simplicity.

* **Fisher Equation ($1 + r_{nominal} = (1 + r_{real})(1 + \pi)$ or its rearranged forms):** The Fisher equation provides the exact relationship and should be used when greater accuracy is required, especially in situations with high inflation rates. In periods of hyperinflation or when dealing with precise calculations, the Fisher equation is essential to avoid significant errors.  For example, if inflation is 20% and the real rate is 5%, the approximate nominal rate is $20\% + 5\% = 25\%$. Using the Fisher equation, $r_{nominal} = (1 + 0.05)(1 + 0.20) - 1 = 1.05 \times 1.20 - 1 = 1.26 - 1 = 0.26 = 26\%$. The difference of 1 percentage point can be significant in many financial contexts.

In summary, while the approximation is useful for simple understanding and quick calculations under low inflation, the Fisher equation provides the accurate relationship between nominal rates, real rates, and inflation and should be used for precise analysis, especially when inflation is not negligible.

The three main types of Treasury securities are Treasury Bills, Treasury Notes, and Treasury Bonds. They are distinguished primarily by their maturity periods.

**1. Treasury Bills (T-bills):** These are short-term securities with maturities of one year or less. Common maturities are 4, 8, 13, 17, 26, and 52 weeks. T-bills are sold at a discount from their face value, and they do not pay periodic interest payments. Instead, the investor's return is the difference between the purchase price and the face value received at maturity. This is known as a discount yield. For example, if an investor purchases a \$1,000 T-bill for \$980, the \$20 difference represents the interest earned when the bill matures and the investor receives \$1,000.

T-notes pay a fixed interest rate, known as the coupon rate, which is paid semiannually until maturity.  At maturity, the investor receives the face value of the note. For instance, a \$1,000 Treasury Note with a 3% coupon rate will pay \$15 in interest every six months (\$1,000 * 0.03 / 2) and will return the \$1,000 principal at the end of the term.

**3. Treasury Bonds (T-bonds):** These are long-term securities with maturities of more than ten years.  Currently, Treasury Bonds are typically issued with 20-year and 30-year maturities. In summary, the key differences are their maturity periods and how they are sold:

*   **Maturity:** T-bills are short-term (less than one year), T-notes are medium-term (two to ten years), and T-bonds are long-term (greater than ten years).
*   **Interest Payments:** T-bills are discount securities with no periodic interest payments, while T-notes and T-bonds pay fixed semiannual interest payments based on their coupon rates.
*   **Sale Mechanism:** T-bills are sold at a discount, while T-notes and T-bonds are typically sold at or near par value and accrue interest based on their coupon rates.

These differences in maturity and interest payment structure cater to various investor preferences and needs regarding investment horizon and income streams. Investors seeking short-term, liquid, and low-risk investments often favor T-bills. Investors with medium to long-term investment horizons and a desire for regular income may prefer T-notes or T-bonds.

**Question 17:** 17. Does a high forward interest rate predict a high future interest rate?

**Answer:** Yes, generally a high forward interest rate tends to predict a high future interest rate, although this is not a perfect predictor due to various factors. This relationship is primarily explained by the Expectations Hypothesis of the term structure of interest rates.

The forward interest rate is the interest rate agreed upon today for a transaction that will occur at a future date. It is embedded in the yield curve and can be derived from the spot rates (current interest rates for immediate transactions). For example, the forward rate for a period between time $t_1$ and $t_2$ can be calculated from the spot rates for time $t_1$ and $t_2$.

The future interest rate, on the other hand, is the spot interest rate that will prevail at some future date. This is unknown today and is what market participants are trying to predict.

The Expectations Hypothesis posits that forward interest rates are unbiased predictors of future spot interest rates. In its simplest form, the hypothesis states that the forward rate for a future period equals the expected future spot rate for that period.

Let $r_{1t}$ be the spot interest rate for a one-period investment starting at time 0 and maturing at time 1. Let $r_{2t}$ be the spot interest rate for a two-period investment starting at time 0 and maturing at time 2. Let $f_{1,t}$ be the forward interest rate agreed at time 0 for a one-period investment starting at time 1 and maturing at time 2.

According to the Expectations Hypothesis, the two-period investment return should be equal to the expected return from investing in two consecutive one-period investments.  This can be expressed mathematically.

For a two-period investment, the return is $(1+r_{2t})^2$.
For two consecutive one-period investments, the expected return is $(1+r_{1t})(1+E[r_{1,t+1}])$, where $E[r_{1,t+1}]$ is the expected one-period spot rate at time 1.

The Expectations Hypothesis suggests these two returns should be approximately equal:
$$(1+r_{2t})^2 \approx (1+r_{1t})(1+E[r_{1,t+1}])$$

From the relationship between spot and forward rates, we know that:
$$(1+r_{2t})^2 = (1+r_{1t})(1+f_{1,t})$$

Combining these two equations, we get:
$$(1+r_{1t})(1+f_{1,t}) \approx (1+r_{1t})(1+E[r_{1,t+1}])$$

Dividing both sides by $(1+r_{1t})$ (assuming $1+r_{1t} \neq 0$), we are left with:
$$1+f_{1,t} \approx 1+E[r_{1,t+1}]$$

Therefore, the forward rate is approximately equal to the expected future spot rate:
$$f_{1,t} \approx E[r_{1,t+1}]$$

This equation shows that the forward interest rate $f_{1,t}$ is an expectation of the future short-term interest rate $r_{1,t+1}$.  If the forward rate $f_{1,t}$ is high, it implies that the expected future spot rate $E[r_{1,t+1}]$ is also high. Thus, a high forward interest rate predicts a high future interest rate, based on the Expectations Hypothesis.

However, the pure Expectations Hypothesis is often considered too simplistic in reality. Market participants may demand a risk premium for holding longer-term bonds. This risk premium arises because longer-term bond prices are more sensitive to interest rate changes, introducing greater interest rate risk.

If we incorporate a term premium, $\theta_{1,t}$, into the relationship, the equation becomes:
$$f_{1,t} = E[r_{1,t+1}] + \theta_{1,t}$$

Here, $\theta_{1,t}$ represents the term premium for the period between time 1 and time 2.  The term premium is typically positive and increases with the term to maturity.

In this case, the forward rate is not solely an expectation of the future spot rate but also includes a risk premium.  Even with a term premium, a high forward rate still generally implies a high expected future spot rate, although it overestimates the expected future spot rate by the amount of the term premium.  However, if the term premium is relatively stable or less variable than the expected future interest rate changes, a high forward rate will still be a reasonable indicator of a high future interest rate.

In summary, while a high forward interest rate generally predicts a high future interest rate as suggested by the Expectations Hypothesis, it is not a perfect predictor.  The relationship is influenced by factors like risk premiums embedded in forward rates. Nevertheless, under the assumption that term premiums are relatively stable, a high forward rate is still a reasonable indication of a high expected future interest rate.

**Question 35:** 35. The 1-year forward interest rates are as follows:

**Answer:** 1-year forward interest rates are interest rates agreed upon today for a loan that will start one year from today and last for one year. These rates are implied by the current term structure of spot interest rates and represent the market's expectation of what the 1-year spot rate will be one year into the future.

To understand how forward rates are determined, consider the principle of no-arbitrage.  If it were possible to consistently earn a risk-free profit by trading between different maturities, arbitrageurs would exploit these opportunities until the prices adjust to eliminate such profits.

Let $r_{0,1}$ be the 1-year spot rate (the interest rate for a loan starting today and lasting for one year).
Let $r_{0,2}$ be the 2-year spot rate (the interest rate for a loan starting today and lasting for two years).
Let $f_{1,1}$ be the 1-year forward rate starting one year from today (the interest rate for a loan starting in one year and lasting for one year).

Consider two investment strategies:

Strategy 1: Invest \$1 today for 2 years at the 2-year spot rate.  At the end of 2 years, you will have $(1 + r_{0,2})^2$.

Strategy 2: Invest \$1 today for 1 year at the 1-year spot rate. At the end of 1 year, you will have $(1 + r_{0,1})$.  Then, at the end of year 1, invest this amount for another year at the 1-year forward rate $f_{1,1}$. At the end of 2 years, you will have $(1 + r_{0,1})(1 + f_{1,1})$.

In an efficient market without arbitrage opportunities, the returns from these two strategies must be equal. Therefore:

$$(1 + r_{0,2})^2 = (1 + r_{0,1})(1 + f_{1,1})$$

We can solve for the 1-year forward rate, $f_{1,1}$:

$$1 + f_{1,1} = \frac{(1 + r_{0,2})^2}{(1 + r_{0,1})}$$

$$f_{1,1} = \frac{(1 + r_{0,2})^2}{(1 + r_{0,1})} - 1$$

This formula shows how the 1-year forward rate starting in one year is derived from the 1-year and 2-year spot rates.  In general, for any two consecutive periods, we can generalize this relationship.  For example, to find the 1-year forward rate starting in year $n-1$ (denoted as $f_{n-1,1}$), given the $n$-year spot rate $r_{0,n}$ and the $(n-1)$-year spot rate $r_{0,n-1}$, we can use a similar no-arbitrage argument:

$$(1 + r_{0,n})^n = (1 + r_{0,n-1})^{n-1} (1 + f_{n-1,1})$$

Solving for $f_{n-1,1}$:

$$1 + f_{n-1,1} = \frac{(1 + r_{0,n})^n}{(1 + r_{0,n-1})^{n-1}}$$

$$f_{n-1,1} = \frac{(1 + r_{0,n})^n}{(1 + r_{0,n-1})^{n-1}} - 1$$

In the question's context, "1-year forward interest rates are as follows" suggests that a series of forward rates for 1-year periods starting at different times in the future are expected to be provided. For instance, we could have:

$f_{0,1}$ (1-year spot rate, also considered as 1-year forward rate starting today)

$f_{3,1}$ (1-year forward rate starting in 3 years)
and so on.

These 1-year forward rates provide insights into the market's expectations about the path of future short-term interest rates.  An upward sloping forward rate curve (where $f_{1,1} > f_{0,1}$, $f_{2,1} > f_{1,1}$, etc.) suggests that the market expects interest rates to rise in the future.  A downward sloping curve indicates expectations of falling interest rates. A flat curve suggests expectations of stable future interest rates.

**Answer:** Yes, generally, long-term bonds tend to pay more than short-term bonds. This difference in yield is primarily due to the **term premium**, also known as the **maturity premium**. The term premium is the additional compensation investors demand for the increased risks associated with holding bonds for longer periods.  The idea that you only get your money back after a long time and might need it earlier is a component of these risks that contribute to the term premium.

Here is a detailed explanation:

**1. Time Value of Money and Opportunity Cost:**

Your intuition about needing money earlier is related to the fundamental concept of the time value of money.  Money received today is generally worth more than the same amount received in the future. This is because money in hand today can be invested to generate returns, or it can be used to meet immediate needs and opportunities.

When you invest in a long-term bond, you are essentially locking up your capital for an extended period. During this time, you forgo the opportunity to access that principal for other purposes.  This opportunity cost is a factor in why investors demand higher yields for long-term bonds. They are being compensated for the reduced flexibility and potential foregone opportunities.

**2. Interest Rate Risk:**

Long-term bonds are significantly more sensitive to changes in interest rates than short-term bonds. This is known as **interest rate risk** or **duration risk**.

*   **Price Sensitivity:** If interest rates rise, the market value of existing bonds will fall.  The longer the maturity of a bond, the greater its price sensitivity to interest rate changes.  This is because the future cash flows of a long-term bond are discounted over a longer period, making their present value more sensitive to changes in the discount rate (interest rate).
*   **Reinvestment Risk (for short-term bonds):** While long-term bonds have higher interest rate risk in terms of price sensitivity, short-term bonds are more exposed to **reinvestment risk**.  If interest rates fall, investors in short-term bonds, upon maturity, will have to reinvest their principal at lower prevailing interest rates, potentially reducing their overall return.

However, the term premium primarily reflects compensation for the *price sensitivity* aspect of interest rate risk, which is more pronounced for long-term bonds. Investors in long-term bonds are exposed to the risk that interest rates might rise significantly over the life of the bond, causing a substantial decline in the bond's market value if they need to sell it before maturity.  To compensate for this heightened risk, long-term bonds typically offer higher yields at issuance.

**3. Inflation Risk:**

Inflation erodes the purchasing power of money over time.  Long-term bonds are exposed to greater **inflation risk** than short-term bonds.

*   **Uncertainty of Future Inflation:** Inflation rates are inherently uncertain, especially over longer horizons.  Investors in long-term bonds face the risk that inflation might be higher than anticipated over the bond's life, reducing the real (inflation-adjusted) return of their investment.
*   **Fixed Nominal Payments:** Most bonds offer fixed nominal coupon payments. If inflation rises unexpectedly, the real value of these fixed payments decreases.  This effect is more significant for long-term bonds because the fixed payments extend further into the future, where the impact of unexpected inflation can accumulate.

To compensate for this inflation risk, investors demand a higher yield on long-term bonds to protect the real value of their returns against potential inflationary pressures over the investment horizon.

**4. Liquidity Risk:**

While government bonds of highly liquid markets are generally very liquid regardless of maturity, in less liquid markets or for corporate bonds, longer-term bonds can sometimes be slightly less liquid than short-term bonds.

*   **Lower Trading Volume:**  Longer-term bonds might sometimes have lower trading volumes compared to more frequently traded short-term issues. This can make it slightly harder to sell a long-term bond quickly at a desired price, especially in stressed market conditions.
*   **Holding Period Uncertainty:** Investors in long-term bonds may have less certainty about their holding period.  Unexpected needs to sell the bond before maturity might be more likely over a longer time frame, and the potential for reduced liquidity could be a concern.

Although liquidity risk is generally less dominant than interest rate risk and inflation risk in determining the term premium for highly rated government bonds, it can still contribute to the higher yields demanded for longer maturities, particularly for less liquid bond markets or corporate issues.

**5. The Yield Curve and Term Premium:**

The relationship between bond yields and their maturities is graphically represented by the **yield curve**.  In a normal economic environment, the yield curve is typically **upward sloping**. This means that yields on bonds increase as their maturity extends.  This upward slope largely reflects the term premium.

Mathematically, the yield of a long-term bond can be thought of as incorporating expectations of future short-term interest rates plus a term premium:

$$ Y_{n} \approx \frac{1}{n} \left( E[r_1] + E[r_2] + \dots + E[r_n] \right) + TP_n $$

Where:

*   $Y_{n}$ is the yield of an n-period bond.
*   $E[r_i]$ is the expected short-term interest rate in period $i$.
*   $TP_n$ is the term premium for an n-period bond.

This equation is a simplified representation, but it highlights that the yield of a long-term bond is not just an average of expected future short-term rates but also includes a term premium ($TP_n$) to compensate for the risks discussed above.

**In summary:**

Long-term bonds generally pay more than short-term bonds primarily due to the **term premium**. This premium compensates investors for:

*   **Interest rate risk:** The greater price sensitivity of long-term bonds to interest rate changes.
*   **Inflation risk:** The increased uncertainty of inflation and its potential impact on the real value of returns over longer periods.
*   **Opportunity cost:** The reduced flexibility and foregone opportunities from locking up capital for a longer duration, which is linked to the idea of needing money earlier.
*   **Liquidity risk (potentially):**  Slightly reduced liquidity in some cases compared to short-term bonds.

The expectation of higher yields on long-term bonds is a fundamental characteristic of bond markets and is reflected in the typically upward-sloping yield curve. While there can be periods where the yield curve flattens or even inverts (short-term yields higher than long-term yields), these situations are usually associated with specific economic conditions and expectations and do not negate the general principle of a term premium compensating for the risks of longer-term investments.

**Question 39:** 39. The yield curve is usually upward sloping. What does this mean?

**Answer:** An upward sloping yield curve signifies that bonds with longer maturities offer higher yields compared to bonds with shorter maturities. This positive slope is commonly observed and is often referred to as a normal yield curve.

Several factors contribute to this typical upward slope:

**1. Expectations of Future Interest Rates:**  A primary driver of the yield curve's shape is the market's expectation of future short-term interest rates.  Longer-term bond yields can be viewed as an average of expected future short-term interest rates plus a term premium.  If investors anticipate that short-term interest rates will rise in the future, they will demand higher yields on longer-term bonds to compensate for the opportunity cost of being locked into a lower rate for an extended period.

Investors demand a higher yield on longer-term bonds to compensate for the uncertainty and potential loss of purchasing power due to inflation over a longer time horizon. Liquidity refers to the ease with which an asset can be bought or sold in the market without significantly affecting its price. Investors may prefer the flexibility and lower price volatility associated with shorter-term bonds. To incentivize investors to hold less liquid, longer-term bonds, a liquidity premium is offered, resulting in higher yields for longer maturities.

**4. Interest Rate Risk (Duration Risk):**  Longer-term bonds are more sensitive to changes in interest rates than shorter-term bonds. This sensitivity is measured by duration.  If interest rates rise, the price of a longer-term bond will fall more significantly than the price of a shorter-term bond.  Investors require a premium for bearing this greater interest rate risk associated with longer maturities.

**5. Risk Premium for Uncertainty:**  The future is inherently uncertain, and this uncertainty increases over longer time horizons. Investors demand a risk premium to compensate for this increased uncertainty associated with longer-term investments.

*   **Economic Growth:**  Anticipated economic growth typically leads to increased demand for credit, potentially pushing interest rates higher in the future.
*   **Stable or Rising Inflation:**  An upward sloping curve can reflect expectations of stable or increasing inflation, requiring higher yields to maintain real returns.
*   **Normal Economic Conditions:**  It is considered the "normal" shape because it reflects the typical compensation demanded by investors for time, risk, and illiquidity.

In contrast, a flat or inverted yield curve (where longer-term yields are lower than shorter-term yields) can signal economic slowdown or recessionary expectations, as investors anticipate future interest rate cuts by central banks in response to weakening economic conditions.

In summary, an upward sloping yield curve is the typical shape, indicating that investors demand higher yields for longer-term bonds to compensate for increased interest rate risk, inflation risk, liquidity preference, and the uncertainty associated with longer time horizons, often reflecting expectations of future economic growth and stable or rising interest rates.

The investor's return comes from the difference between the purchase price and the face value received at maturity.

Under normal economic conditions, the yield curve is **upward sloping**. This means that yields on longer-term securities are higher than yields on shorter-term securities. This upward slope reflects the **term premium** or **maturity premium**, which is the additional compensation investors demand for bearing the greater risks associated with longer-term bonds.

It is important to note that under certain economic conditions, such as during periods of economic recession or when the central bank is expected to aggressively lower interest rates, the yield curve can become flat or even inverted (downward sloping). In an inverted yield curve, short-term interest rates are higher than long-term interest rates. However, under normal circumstances, the statement that investors earn a higher annualized rate of return from long-term T-bonds than short-term T-bills holds true due to the risk-return trade-off and the typical upward slope of the yield curve.

**Question 41:** 41. Does the evidence suggest that long-term bonds tend to earn higher average rates of return than short-term bonds? If yes, why is this the case? If no, why is this not possible?

**Answer:** Yes, evidence generally suggests that long-term bonds tend to earn higher average rates of return than short-term bonds. This phenomenon is primarily attributed to the **term premium**, which is the additional compensation investors demand for bearing the increased risks associated with investing in long-term bonds compared to short-term bonds.

The rationale behind this observation can be explained through the following points:

1.  **Interest Rate Risk:** Long-term bonds are more sensitive to changes in interest rates than short-term bonds. This is because a larger portion of a long-term bond's value is derived from cash flows further into the future, and the present value of these distant cash flows is more significantly affected by changes in discount rates (interest rates).

However, the price of the 20-year bond will fall by a greater percentage than the price of the 1-year bond. This is because the discounted value of the distant cash flows (coupons and principal repayment in 20 years) is more sensitive to the change in interest rates.  This price sensitivity is often referred to as **duration**. Long-term bonds have higher duration, meaning they are more exposed to interest rate risk.

2.  **Reinvestment Risk:** While long-term bonds are more exposed to interest rate risk, short-term bonds are more susceptible to reinvestment risk. Reinvestment risk is the risk that future interest rates will be lower than the current rate when the short-term bond matures and the proceeds need to be reinvested.

    Investors in short-term bonds face the uncertainty of reinvesting their principal at potentially lower rates when the bonds mature.  In contrast, investors in long-term bonds lock in a yield for a longer period, reducing the frequency of reinvestment decisions and therefore mitigating reinvestment risk to some extent over the long term. 3.  **Inflation Risk:** Long-term bonds are also exposed to greater inflation risk. Inflation erodes the real value of future cash flows. Investors in long-term bonds face the risk that inflation will unexpectedly increase, reducing the real return on their investment.  Short-term bonds, maturing sooner, are less exposed to long-term inflationary uncertainties.

This term premium is the extra return investors expect to receive for holding a long-term bond compared to a short-term bond. This premium is reflected in the yield curve, which is typically upward sloping, meaning that yields on longer-term bonds are generally higher than yields on shorter-term bonds.

    The expected return on a long-term bond can be thought of as comprising two components: the expected return on a short-term bond plus a term premium.

    $$ \text{Expected Return}_{\text{Long-term}} = \text{Expected Return}_{\text{Short-term}} + \text{Term Premium} $$

    The term premium is not constant and can vary over time depending on economic conditions and investor sentiment regarding risk and uncertainty about future interest rates and inflation.

5.  **Empirical Evidence:** Historical data from bond markets in various countries generally supports the notion that long-term bonds have, on average, yielded higher returns than short-term bonds over extended periods. This historical outperformance is considered to be the empirical realization of the term premium. However, it is important to note that in specific periods, especially during times of economic uncertainty or when central banks aggressively lower interest rates, short-term bonds may temporarily outperform long-term bonds. This is because in such scenarios, the expectation of falling interest rates may cause long-term bond prices to rise significantly, but this is often a temporary phenomenon and does not negate the long-term tendency for long-term bonds to offer higher average returns due to the term premium.

In summary, the evidence suggests that long-term bonds tend to earn higher average rates of return than short-term bonds because investors demand a term premium to compensate for the greater interest rate risk, inflation risk, and overall uncertainty associated with longer maturities. This term premium is a fundamental aspect of bond pricing and the term structure of interest rates, reflecting the risk-return trade-off in fixed income markets.

**Question 42:** 42. If you want to commit to saving at an interest rate of $f_{3, 4}$ in December 2004, what would you have to do? (Assume any amount of investment you wish, and work from there.)

**Answer:** To commit to saving at an interest rate of $f_{3, 4}$ in December 2004, you would need to lock in this forward rate for an investment period starting in December 2007 and ending in December 2008. The forward rate $f_{3, 4}$ is derived from the spot rates for 3 and 4 years in December 2004.

Let $r_3$ be the spot interest rate for a 3-year investment and $r_4$ be the spot interest rate for a 4-year investment, both prevailing in December 2004. The forward rate $f_{3, 4}$ is the implied interest rate for a one-year period starting 3 years from December 2004 (i.e., starting in December 2007) and ending 4 years from December 2004 (i.e., ending in December 2008).

The relationship between spot rates and forward rates is based on the principle of no arbitrage. Investing for 4 years at the spot rate $r_4$ should yield the same total return as investing for 3 years at the spot rate $r_3$ and then reinvesting the proceeds for one more year at the forward rate $f_{3, 4}$. Mathematically, this can be expressed as:

$(1 + r_4)^4 = (1 + r_3)^3 (1 + f_{3, 4})$

To find the forward rate $f_{3, 4}$, we rearrange the formula:

$1 + f_{3, 4} = \frac{(1 + r_4)^4}{(1 + r_3)^3}$

$f_{3, 4} = \frac{(1 + r_4)^4}{(1 + r_3)^3} - 1$

To commit to saving at this forward rate $f_{3, 4}$, you would essentially need to enter into a Forward Rate Agreement (FRA) or construct a portfolio using spot market instruments that replicates the desired forward rate.  In practice, the most direct way to commit to saving at the rate $f_{3, 4}$ is to enter into an FRA.

Here's what you would do conceptually:

1.  **Determine the spot rates $r_3$ and $r_4$ in December 2004.** These rates are observable in the market for zero-coupon bonds or can be derived from the yields of coupon-bearing bonds.

2.  **Calculate the forward rate $f_{3, 4}$ using the formula:**
    $$f_{3, 4} = \frac{(1 + r_4)^4}{(1 + r_3)^3} - 1$$

3.  **Enter into a Forward Rate Agreement (FRA).**  You would agree with a financial institution (the counterparty) to lend a notional amount starting in 3 years for a period of 1 year at the rate $f_{3, 4}$.  The FRA specifies:
    *   The forward rate: $f_{3, 4}$
    *   The notional principal amount (you can choose any amount, let's say \$1,000,000 for illustration)
    *   The future period: starting in 3 years and ending in 4 years from December 2004.

In an FRA, no principal is actually exchanged. At the beginning of the forward period (in December 2007), the prevailing spot rate for a 1-year loan starting at that time (let's call it $r_{spot}$) is compared with the agreed forward rate $f_{3, 4}$.

*   If $r_{spot} > f_{3, 4}$, the counterparty pays you the interest rate differential, effectively compensating you for the difference between the actual spot rate and the locked-in forward rate.
*   If $r_{spot} < f_{3, 4}$, you pay the counterparty the interest rate differential.

The payoff at year 3 (December 2007), discounted back to year 3, is proportional to $(r_{spot} - f_{3, 4})$.  To commit to *saving* at the rate $f_{3, 4}$, you would enter into an FRA as the lender, ensuring you receive interest based on the rate $f_{3, 4}$ for the period between year 3 and year 4, regardless of the actual spot rate at year 3.

**Example:**

Assume in December 2004, the spot rates are:
$r_3 = 0.03$ (3%)
$r_4 = 0.04$ (4%)

Calculate $f_{3, 4}$:
$$f_{3, 4} = \frac{(1 + 0.04)^4}{(1 + 0.03)^3} - 1$$
$$f_{3, 4} = \frac{(1.04)^4}{(1.03)^3} - 1$$
$$f_{3, 4} = \frac{1.16985856}{1.092727} - 1$$
$$f_{3, 4} \approx 1.07058 - 1$$
$$f_{3, 4} \approx 0.07058$$
$$f_{3, 4} \approx 7.058\%$$

To commit to saving at approximately $7.058\%$ for the period between December 2007 and December 2008, you would enter into an FRA in December 2004 to receive interest at the rate of $7.058\%$ on a notional principal for this future period. This action locks in the forward rate $f_{3, 4}$ for your savings, providing certainty about the interest rate you will effectively earn for that future investment period.

- Year 1: Interest payment of $1,000
- Year 2: Interest payment of $1,000 plus principal repayment of $25,000, totaling $26,000

$25,000 = \frac{1,000}{(1+y)^1} + \frac{26,000}{(1+y)^2}$

$25,000 = \frac{1,000}{x} + \frac{26,000}{x^2}$

Multiply both sides by $x^2$ to eliminate the denominators:

$25,000x^2 = 1,000x + 26,000$

Rearrange the equation into a quadratic form:

$25,000x^2 - 1,000x - 26,000 = 0$

Divide the entire equation by 1,000 to simplify:

$25x^2 - x - 26 = 0$

Use the quadratic formula to solve for $x$:

$x = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a}$

$x = \frac{-(-1) \pm \sqrt{(-1)^2 - 4(25)(-26)}}{2(25)}$

$x = \frac{1 \pm 51}{50}$

**Question 45:** 45. A bond pays $150 for every $100 invested. What is its continuously compounded interest rate?

**Answer:** The continuously compounded interest rate represents the theoretical interest rate that would accrue if interest were constantly being added to the principal, with reinvestment of the interest itself also earning interest immediately. The formula for calculating the future value of an investment with continuously compounded interest is given by:
$$A = Pe^{rt}$$
where:
$A$ is the future value of the investment,
$P$ is the principal amount (initial investment),
$r$ is the continuously compounded interest rate (expressed as a decimal),
$t$ is the time period (in years),
$e$ is the base of the natural logarithm, approximately equal to 2.71828.

In this problem, we are given that for every $100 invested, the bond pays $150. This means that the initial investment $P = 100$ grows to a future value $A = 150$.  Although the time period is not explicitly stated, it is standard practice in such problems to assume the time period to be one year ($t=1$) unless otherwise specified.  We need to find the continuously compounded interest rate $r$.

Substituting the given values into the formula, we get:
$$150 = 100e^{r \times 1}$$
$$150 = 100e^{r}$$

To solve for $r$, we first isolate the exponential term by dividing both sides of the equation by 100:
$$\frac{150}{100} = e^{r}$$
$$1.5 = e^{r}$$

To solve for $r$, we take the natural logarithm (ln) of both sides of the equation. The natural logarithm is the inverse function of the exponential function with base $e$, so $\ln(e^x) = x$.
$$\ln(1.5) = \ln(e^{r})$$
$$\ln(1.5) = r$$

Now we calculate the numerical value of $\ln(1.5)$. Using a calculator, we find:
$$r \approx 0.405465$$

To express the interest rate as a percentage, we multiply by 100:
$$r \approx 0.405465 \times 100\% = 40.5465\%$$

Therefore, the continuously compounded interest rate is approximately 40.5465%.  Rounding to two decimal places, we get 40.55%.

Final Answer: The final answer is $\boxed{40.55\%}$

**Question 46:** 46. Confirm my claim that you can add continuously compounded interest rates. That is, a bond pays a continuously compounded interest rate of 10%. Upon maturity, the money can be reinvested at a continuously compounded interest rate of 20%. If you invest $100 today, how much money will you end up with? What is the simple and continuously compounded interest rate over the 2 periods?

**Answer:** The final amount of money after the two investments is calculated as follows:

First investment period:
Initial investment, $P = \$100$
Continuously compounded interest rate, $r_1 = 10\% = 0.10$
Assuming a period of 1 year for the first investment, $t_1 = 1$ year.
The amount after the first period, $A_1$, is given by the continuous compounding formula:
$A_1 = P e^{r_1 t_1} = 100 e^{0.10 \times 1} = 100 e^{0.10}$
$A_1 \approx 100 \times 1.10517 = \$110.517$

Second investment period:
Principal for the second period is the amount from the first period, $P_2 = A_1 = \$110.517$
Continuously compounded interest rate, $r_2 = 20\% = 0.20$
Assuming a period of 1 year for the second investment, $t_2 = 1$ year.
The amount after the second period, $A_2$, is given by the continuous compounding formula:
$A_2 = P_2 e^{r_2 t_2} = 110.517 e^{0.20 \times 1} = 110.517 e^{0.20}$
$A_2 \approx 110.517 \times 1.22140 \approx \$135.049$

The final amount after the two periods is approximately $ \$135.05$.

To calculate the simple interest rate over the 2 periods, we consider the total interest earned and the total time.
Total interest earned = $A_2 - P = \$135.049 - \$100 = \$35.049$
Total time period, $T = t_1 + t_2 = 1 + 1 = 2$ years.
Simple interest rate, $r_{simple}$, is calculated using the formula:
$Interest = P \times r_{simple} \times T$
$35.049 = 100 \times r_{simple} \times 2$
$r_{simple} = \frac{35.049}{100 \times 2} = \frac{35.049}{200} = 0.175245$
The simple interest rate over the 2 periods is approximately $17.52\%$.

To calculate the continuously compounded interest rate over the 2 periods, we use the formula:
$A_2 = P e^{r_{cc} T}$
$135.049 = 100 e^{r_{cc} \times 2}$
$\frac{135.049}{100} = e^{2r_{cc}}$
$1.35049 = e^{2r_{cc}}$
Taking the natural logarithm of both sides:
$\ln(1.35049) = 2r_{cc}$
$r_{cc} = \frac{\ln(1.35049)}{2}$
$r_{cc} \approx \frac{0.30010}{2} = 0.15005$
The continuously compounded interest rate over the 2 periods is approximately $15.01\%$.

Continuously compounded interest rates cannot be directly added in the way that simple interest rates might sometimes be approximated. When interest is compounded continuously, the growth factor for each period is multiplicative, not additive in terms of the rates. For the first period, the growth factor is $e^{r_1 t_1}$, and for the second period, it is $e^{r_2 t_2}$. The total growth factor over both periods is the product of the individual growth factors:
$Growth_{total} = e^{r_1 t_1} \times e^{r_2 t_2} = e^{r_1 t_1 + r_2 t_2} = e^{(r_1 t_1 + r_2 t_2)}$
If one were to incorrectly "add" the continuously compounded interest rates directly to get an overall rate, it might be interpreted as summing the rates and applying that sum over the total time.  For example, if one were to add the rates in this scenario to get $10\% + 20\% = 30\%$ and apply this over 2 years, the calculation would be $A = 100 e^{(0.30) \times 2} = 100 e^{0.60} \approx \$182.21$. This is significantly different from the actual amount of $ \$135.05$.

What is actually additive in the case of consecutive continuously compounded interest periods is the product of the growth factors, or equivalently, the sum of the exponents in the exponential growth functions, which are $r_1 t_1$ and $r_2 t_2$. To find a single equivalent continuously compounded rate ($r_{cc}$) over the total time $T = t_1 + t_2$, we equate the total growth factor:
$e^{r_{cc} T} = e^{r_1 t_1 + r_2 t_2}$
This implies $r_{cc} T = r_1 t_1 + r_2 t_2$, so $r_{cc} = \frac{r_1 t_1 + r_2 t_2}{T}$. In the case where $t_1 = t_2 = 1$, $T = 2$, then $r_{cc} = \frac{r_1 + r_2}{2}$, which is the average of the rates, not the sum.

In summary, while you cannot directly add continuously compounded interest rates to find a simple overall rate for consecutive periods, the total accumulated amount is found by sequentially applying the continuous compounding formula for each period. The overall continuously compounded rate over multiple periods is related to the weighted average of the rates, weighted by the duration of each period.

Final Answer:
The final amount you will end up with is approximately $ \$135.05$. The simple interest rate over the 2 periods is approximately $17.52\%$. The continuously compounded interest rate over the 2 periods is approximately $15.01\%$. Continuously compounded interest rates cannot be directly added to find an overall rate in the context of consecutive investment periods. Instead, the growth factors associated with each continuously compounded rate are multiplied to find the total growth factor over multiple periods. The overall continuously compounded rate is effectively the time-weighted average of the individual rates when considering consecutive periods.

**Question 47:** 47. Assume the 12-month Treasury note costs $10,065.22 and pays a coupon of $150 in 6 months, and interest plus coupon of $10,150 in 12 months. Now assume the 6-month Treasury bill costs $10,103.96 and has only one remaining coupon-plus-principal payment of $10,200. **Answer:** The yield of the 12-month Treasury note can be calculated by setting the present value of its future cash flows equal to its current cost. The cash flows are a \$150 coupon payment in 6 months and a \$10,150 payment (comprising the final coupon and principal) in 12 months.  Let $y_{note}$ be the annualized yield of the 12-month note, and we assume semiannual compounding. The present value equation is:

$$10065.22 = \frac{150}{(1 + y_{note}/2)^1} + \frac{10150}{(1 + y_{note}/2)^2}$$

To solve for $y_{note}$, let $r = y_{note}/2$. The equation becomes:

$$10065.22 = \frac{150}{(1 + r)} + \frac{10150}{(1 + r)^2}$$

Multiplying both sides by $(1+r)^2$ to eliminate denominators:

$$10065.22 (1 + r)^2 = 150(1 + r) + 10150$$

Expanding the equation:

$$10065.22 (1 + 2r + r^2) = 150 + 150r + 10150$$
$$10065.22 + 20130.44r + 10065.22r^2 = 10300 + 150r$$

Rearranging into a quadratic equation in the form $ar^2 + br + c = 0$:

$$10065.22r^2 + (20130.44 - 150)r + (10065.22 - 10300) = 0$$
$$10065.22r^2 + 19980.44r - 234.78 = 0$$

Using the quadratic formula, $r = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a}$:

$$r = \frac{-19980.44 \pm \sqrt{(19980.44)^2 - 4(10065.22)(-234.78)}}{2(10065.22)}$$
$$r = \frac{-19980.44 \pm \sqrt{399218009.59 + 9462598.7}}{20130.44}$$
$$r = \frac{-19980.44 \pm \sqrt{408680608.29}}{20130.44}$$
$$r = \frac{-19980.44 \pm 20215.85}{20130.44}$$

Since yield must be positive, we take the positive root:

$$r = \frac{-19980.44 + 20215.85}{20130.44} = \frac{235.41}{20130.44} \approx 0.011694$$

The annualized yield for the 12-month note is $y_{note} = 2r = 2 \times 0.011694 \approx 0.023388$ or 2.3388%.
